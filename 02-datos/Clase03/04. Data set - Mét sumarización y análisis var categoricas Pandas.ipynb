## Slides 10–11 – Ejemplo Aplicado (todos los métodos numéricos)

```python
# ── Slides 10–11: Ejemplo aplicado completo ───────────────────────────────────

import pandas as pd
datos = {'Edad': [25, 32, 41, 38, None, 29, 44],
         'Departamento': ['TI', 'TI', 'RRHH', 'TI', 'RRHH', 'Ventas', 'RRHH']}

df = pd.DataFrame(datos)
print("Edad mínima:", df['Edad'].min())
print("Edad máxima:", df['Edad'].max())
print("Promedio de edad:", df['Edad'].mean())
print("Mediana de edad:", df['Edad'].median())
print("Total de registros válidos:", df['Edad'].count())
# Estos métodos se pueden aplicar también a todo el DataFrame:
df[['Edad']].mean()
```

```python
# ── Slide 17: Ejemplo aplicado — análisis de regiones ────────────────────────

df = pd.DataFrame({'Región': ['Norte', 'Sur', 'Centro', 'Norte', 'Sur', 'Centro', 'Centro', 'Norte']})
print("Regiones únicas:", df['Región'].unique())
print("Frecuencia por región:")
print(df['Región'].value_counts())
```

## Slides 19–23 – Ejercicio Guiado: Postulantes a Cargos Tecnológicos

**Objetivo (Slide 20):** Resumen numérico y categórico del DataFrame de postulantes.

```python
# ── Slide 21: Crear el DataFrame de postulantes ───────────────────────────────

import pandas as pd
data = {'Nombre': ['Valentina', 'Pedro', 'Laura', 'Javiera', 'Carlos', 'Martín'],
        'Edad': [29, 35, 31, 28, 42, 38],
        'Experiencia (años)': [3, 8, 5, 2, 10, 6],
        'Departamento': ['TI', 'TI', 'RRHH', 'TI', 'RRHH', 'Ventas']}
df = pd.DataFrame(data)
```

```python
# ── Slide 22: Aplicar todos los métodos ──────────────────────────────────────

print("Edad mínima:", df['Edad'].min())
print("Edad máxima:", df['Edad'].max())
print("Edad promedio:", df['Edad'].mean())
print("Edad mediana:", df['Edad'].median())
print("Total de años de experiencia:", df['Experiencia (años)'].sum())
print("Cantidad de datos válidos:", df['Edad'].count())
print("Departamentos únicos:", df['Departamento'].unique())
print("Frecuencia de departamentos:")
print(df['Departamento'].value_counts())
```

```python
# ── Slide 23: Preguntas de análisis ──────────────────────────────────────────

rango_edad = df['Edad'].max() - df['Edad'].min()
depto_mas_demandado = df['Departamento'].value_counts().idxmax()

print(f'Rango de edad: {rango_edad} años ({df["Edad"].min()} – {df["Edad"].max()})')
print(f'Departamento con mayor demanda: {depto_mas_demandado} ({df["Departamento"].value_counts().max()} postulantes)')
```
