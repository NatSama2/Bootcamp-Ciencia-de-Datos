## Slides 5–6 – Exploración Inicial con `head()` y `tail()`

```python
# ── Slide 5: head() — primeras filas ─────────────────────────────────────────

df.head()    # Primeras 5 filas
df.head(10)  # Primeras 10 filas
```

## Slides 8–9 – Revisión Estructural con `info()`

`info()` entrega:
- Cantidad total de filas y columnas
- Nombres de columnas
- Cantidad de **valores no nulos** por columna
- **Tipo de dato** de cada columna
- Memoria utilizada

```python
# ── Slide 8: info() ───────────────────────────────────────────────────────────

df.info()
```

```python
# ── Slide 9: Ejemplo aplicado ────────────────────────────────────────────────

df = pd.read_csv('viajes_interurbanos.csv')
df.info()
```

---
## Slides 12–13 – Análisis Estadístico Preliminar con `describe()`

`describe()` calcula automáticamente: `count`, `mean`, `std`, `min`, `25%`, `50%`, `75%`, `max`.

```python
# ── Slide 12: describe() ─────────────────────────────────────────────────────

df.describe()
```

```python
# ── Slide 13: Ejemplo aplicado ───────────────────────────────────────────────

df = pd.read_csv('viajes_interurbanos.csv')
df.describe()
```

```python
# ── Slide 13: Para columnas no numéricas ─────────────────────────────────────

df.describe(include='object')   # para texto
df.describe(include='all')      # para todo tipo de datos
```

---
## Slide 16 – Estrategia de Exploración Integrada

Buena práctica profesional: combinar todos los métodos en una **rutina de revisión inicial**.

```python
# ── Slide 16: Rutina de diagnóstico inicial ───────────────────────────────────

print("Vista inicial:")
print(df.head())
print("\nResumen de estructura:")
print(df.info())
print("\nEstadísticas básicas:")
print(df.describe())
```

## Slides 17–20 – Ejercicio Guiado: Cadena de Farmacias

**Objetivo (Slide 18):** Aplicar `head`, `tail`, `info`, `describe` para entregar un resumen a la jefatura.

```python
# ── Slide 19: Crear el DataFrame de ventas de medicamentos ───────────────────

import pandas as pd
datos = {'Producto': ['Paracetamol', 'Ibuprofeno', 'Amoxicilina', 'Aspirina', 'Loratadina'],
         'Precio': [1200, 1500, 2500, 900, 1800],
         'Stock': [35, 20, 0, 50, 40],
         'Farmacia': ['Central', 'Norte', 'Sur', 'Central', 'Oriente']}
df = pd.DataFrame(datos)
```

```python
# ── Slide 20: Ejecutar y analizar los 4 métodos ───────────────────────────────

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
# ── Slide 20: Preguntas de análisis — respuestas con código ──────────────────

print('¿Qué tipo de variables contiene el DataFrame?')
print(df.dtypes)
print()
print('¿Hay alguna columna con valores faltantes?')
print(df.isnull().sum())
print()
print(f'Producto más caro:   {df.loc[df["Precio"].idxmax(), "Producto"]} (${df["Precio"].max():,})')
print(f'Producto más barato: {df.loc[df["Precio"].idxmin(), "Producto"]} (${df["Precio"].min():,})')
print()
print('Distribución del stock:')
print(df['Stock'].describe().round(1))
```

## Slides 21–24 – Ejercicio Guiado: Asistencia a Capacitaciones

**Objetivo (Slide 22):** Explorar y diagnosticar un archivo de asistencia.

```python
# ── Slide 23: Crear el DataFrame de asistencia ───────────────────────────────

data = {'Nombre': ['Luis', 'Andrea', 'Camila', 'Jorge', 'María'],
        'Duración (hrs)': [2, 3, 4, 2, 3],
        'Modalidad': ['Online', 'Presencial', 'Online', 'Presencial', 'Online'],
        'Región': ['RM', 'Biobío', 'RM', 'Valparaíso', 'Araucanía']}
df = pd.DataFrame(data)
```

```python
# ── Slide 24: Aplicar los 4 métodos ──────────────────────────────────────────

df.head()
df.tail()
df.info()
df.describe()
```

```python
# ── Slide 24: Preguntas de reflexión — respuestas con código ─────────────────

print('Variables categóricas:')
print(df.select_dtypes(include='object').columns.tolist())
print()
print('Valores faltantes por columna:')
print(df.isnull().sum())
print()
print(f'Duración promedio: {df["Duración (hrs)"].mean()} hrs')
print()
print('Modalidad más frecuente:')
print(df['Modalidad'].value_counts())
```
