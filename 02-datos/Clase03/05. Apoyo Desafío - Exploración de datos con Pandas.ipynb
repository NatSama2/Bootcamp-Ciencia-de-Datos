## Creación del DataFrame

Copia el siguiente diccionario y úsalo para crear el DataFrame de pandas.

```python
import pandas as pd

datos_ventas = {
    'Producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Webcam', 'Audífonos', 'Laptop', 'Mouse', 'Mouse'],
    'Categoría': ['Electrónica', 'Accesorios', 'Accesorios', 'Electrónica', 'Accesorios', 'Accesorios', 'Electrónica', 'Accesorios', 'Accesorios'],
    'Precio': [1200, 25, 45, 300, 50, 75, 1250, 27, 26],
    'Unidades_Vendidas': [5, 30, 25, 10, 20, 15, 3, 40, 35]
}

df_ventas = pd.DataFrame(datos_ventas)
print(df_ventas)
```

---
## Requerimiento 1 – Exploración Inicial *(4 puntos)*

```python
# Resumen conciso: tipos de datos y valores no nulos
df_ventas.info()
```

```python
# Primeras 5 filas para inspeccionar la estructura
df_ventas.head()
```

```python
# Resumen estadístico descriptivo para columnas numéricas
df_ventas.describe()
```

---
## Requerimiento 2 – Selección y Filtrado *(3 puntos)*

```python
# Columna Producto como Serie de pandas
df_ventas['Producto']
```

```python
# Columnas Producto y Precio
df_ventas[['Producto', 'Precio']]
```

```python
# Ventas donde el Precio fue superior a 100
df_ventas[df_ventas['Precio'] > 100]
```

---
## Requerimiento 3 – Sumarización de Datos *(3 puntos)*

```python
# Precio promedio de todos los productos
precio_promedio = df_ventas['Precio'].mean()
print(f'Precio promedio: ${precio_promedio:.2f}')
```

```python
# Total de unidades vendidas en categoría "Accesorios"
total_accesorios = df_ventas[df_ventas['Categoría'] == 'Accesorios']['Unidades_Vendidas'].sum()
print(f'Total unidades vendidas en Accesorios: {total_accesorios}')
```

```python
# Cuántas veces aparece cada producto en el listado
df_ventas['Producto'].value_counts()
```
