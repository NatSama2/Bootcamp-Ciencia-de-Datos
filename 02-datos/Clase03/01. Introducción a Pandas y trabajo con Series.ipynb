## Slides 8–9 – Características del Tipo de Dato Serie

Una **Serie** es un arreglo unidimensional con etiquetas (índices). Combina la simplicidad de una lista con la flexibilidad de un diccionario.

Útil para: ventas diarias, temperaturas horarias, resultados de evaluaciones, etc.

```python
# ── Slide 9: Estructura de una Serie ─────────────────────────────────────────

import pandas as pd
s = pd.Series([10, 20, 30], index=['lunes', 'martes', 'miércoles'])
print(s)
```

---
## Slide 10 – Creación de una Serie

pandas permite construir Series desde distintas fuentes:

```python
# ── Slide 10: Desde una lista simple ─────────────────────────────────────────
# Esto genera una Serie con índices automáticos: 0, 1, 2

s = pd.Series([5, 8, 12])
print(s)
```

```python
# ── Slide 10: Desde un diccionario ───────────────────────────────────────────
# Las claves se usan como índices, aportando significado contextual

temperaturas = {
    'lun': 4.3,
    'mar': 3.8,
    'mié': 5.0
}
s = pd.Series(temperaturas)
print(s)
```

```python
# ── Slide 10: Especificando el índice manualmente ────────────────────────────
# Ideal cuando trabajamos con registros de sucursales, regiones, etc.

s = pd.Series([100, 200, 300], index=['Sucursal A', 'Sucursal B', 'Sucursal C'])
print(s)
```

---
## Slide 11 – Obtención de Datos de una Serie

```python
# ── Slide 11: Por índice posicional ──────────────────────────────────────────

print(s[0])
# primer elemento
```

```python
# ── Slide 11: Por índice por etiqueta ────────────────────────────────────────

print(s['Sucursal A'])
```

```python
# ── Slide 11: Por segmento (slicing) ─────────────────────────────────────────

print(s[1:3])
# del segundo al tercer elemento
```

---
## Slide 12 – Exploración de Propiedades y Metainformación

```python
# ── Slide 12: Propiedades de una Serie ───────────────────────────────────────

print(s.values)   # valores contenidos
print(s.index)    # etiquetas
print(s.dtype)    # tipo de dato almacenado
```

---
## Slide 13 – Operaciones sobre una Serie

Las Series permiten **operaciones vectorizadas**: se aplican a todos los elementos sin necesidad de bucles.

```python
# ── Slide 13: Operaciones aritméticas ────────────────────────────────────────

suma        = s + 10
multiplicado = s * 2

print('Serie original:')
print(s)
print('\nSerie + 10:')
print(suma)
print('\nSerie × 2:')
print(multiplicado)
```

```python
# ── Slide 13: Condiciones booleanas y filtrado ────────────────────────────────

mayores = s > 100
print(s[mayores])
```

```python
# ── Slide 13: Estadísticas básicas de resumen ─────────────────────────────────

s.mean()      # promedios
max()         # valor máximo
min()         # valor mínimo
s.describe()  # resumen estadístico general
```

```python
# ── Slide 13: Ejemplo aplicado — detección de valores en riesgo ───────────────
# Detectamos condiciones anómalas: temperaturas peligrosas

# Recreamos la serie de temperaturas del ejemplo del slide
s_riesgo = pd.Series([4.3, 3.8, 1.5, 5.0, 1.2, 4.9], 
                     index=['lun', 'mar', 'mié', 'jue', 'vie', 'sáb'])

riesgo = s_riesgo[s_riesgo < 2.0]
print(riesgo)
```

## Actividad Guiada – Análisis completo del desafío inicial

Aplicamos todos los conceptos de la clase al caso farmacéutico del **Slide 3**.

**Archivo:** `temperaturas_refrigeradores.csv`

**Rango aceptable:** 2.0°C – 8.0°C

```python
# ── Guiada, Paso 1: Cargar datos y crear la Serie ────────────────────────────

import pandas as pd

df = pd.read_csv('temperaturas_refrigeradores.csv')

# Creamos la Serie usando la columna de temperaturas
# Usamos el índice del DataFrame como referencia
s_temps = pd.Series(df['temperatura_c'].values, index=df['timestamp'])

print('=== Serie de temperaturas ===')
print(f'Tipo: {type(s_temps)}')
print(f'Longitud: {len(s_temps)}')
print(f'Dtype: {s_temps.dtype}')
print()
print(s_temps.head(10))
```

```python
# ── Guiada, Paso 2: Inspección con propiedades (Slide 12) ────────────────────

print('Valores (primeros 5):', s_temps.values[:5])
print('Índice (primeros 5):', s_temps.index[:5].tolist())
print('Dtype:', s_temps.dtype)
print()
print('=== Estadísticas descriptivas ===')
print(s_temps.describe().round(2))
```

```python
# ── Guiada, Paso 3: Acceso a elementos (Slide 11) ────────────────────────────

print('Primer registro (índice posicional 0):')
print(s_temps.iloc[0])
print()
print('Primeros 5 registros (slicing):')
print(s_temps.iloc[0:5])
print()
print('Acceso por etiqueta (timestamp específico):')
print(s_temps.iloc[10])   # elemento 10
```

```python
# ── Guiada, Paso 4: Operaciones aritméticas (Slide 13) ───────────────────────

# Convertir de Celsius a Fahrenheit: F = C * 9/5 + 32
s_fahrenheit = s_temps * 9/5 + 32

print('Primeras 5 temperaturas en Celsius y Fahrenheit:')
comparacion = pd.DataFrame({
    'Celsius'   : s_temps.values[:5],
    'Fahrenheit': s_fahrenheit.values[:5].round(1)
})
print(comparacion.to_string(index=False))
```

```python
# ── Guiada, Paso 5: Filtrado condicional — detección de anomalías (Slide 13) ──

# Rango aceptable: 2.0°C – 8.0°C
fuera_rango = s_temps[(s_temps < 2.0) | (s_temps > 8.0)]

print(f'Total de registros:         {len(s_temps)}')
print(f'Registros fuera de rango:   {len(fuera_rango)}')
print(f'Porcentaje de anomalías:    {len(fuera_rango)/len(s_temps)*100:.1f}%')
print()
print('=== Registros con temperatura fuera de rango ===')
print(fuera_rango.to_string())
```

```python
# ── Guiada, Paso 6: Visualización básica ─────────────────────────────────────

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Histograma de temperaturas
axes[0].hist(s_temps, bins=25, color='steelblue', edgecolor='white', alpha=0.85)
axes[0].axvline(2.0, color='red', linestyle='--', linewidth=2, label='Límite inferior (2°C)')
axes[0].axvline(8.0, color='red', linestyle='--', linewidth=2, label='Límite superior (8°C)')
axes[0].axvline(s_temps.mean(), color='orange', linestyle='-', linewidth=2,
                label=f'Media = {s_temps.mean():.2f}°C')
axes[0].set_title('Distribución de Temperaturas', fontsize=12)
axes[0].set_xlabel('Temperatura (°C)')
axes[0].set_ylabel('Frecuencia')
axes[0].legend(fontsize=8)

# Clasificación: dentro/fuera de rango
dentro = len(s_temps[(s_temps >= 2.0) & (s_temps <= 8.0)])
fuera  = len(s_temps[(s_temps < 2.0) | (s_temps > 8.0)])
axes[1].bar(['Dentro de rango\n(2–8°C)', 'Fuera de rango'],
            [dentro, fuera],
            color=['#27AE60', '#E74C3C'], alpha=0.85, edgecolor='white')
axes[1].set_title('Clasificación de Registros', fontsize=12)
axes[1].set_ylabel('Cantidad de registros')
for i, v in enumerate([dentro, fuera]):
    axes[1].text(i, v + 0.5, str(v), ha='center', fontweight='bold', fontsize=12)

plt.suptitle('Control de Calidad — Temperaturas de Refrigeradores', fontsize=13)
plt.tight_layout()
plt.show()
```
