## Slide 8 – Creación de un DataFrame desde un Diccionario

```python
# ── Slide 8: Desde un diccionario de listas o arrays ─────────────────────────

import pandas as pd
data = {'Ciudad': ['Santiago', 'Valparaíso', 'Concepción'],
        'Producto': ['Libros', 'Ropa', 'Calzado'],
        'Costo': [4500, 6700, 8900],
        'Entregado': ['Sí', 'No', 'Sí']}
df = pd.DataFrame(data)
print(df)
```

---
## Slide 9 – Más Formas de Crear DataFrames

```python
# ── Slide 9: Desde una lista de diccionarios (cada dict es una fila) ──────────

registros = [{'Ciudad': 'Temuco',  'Producto': 'Ropa',   'Costo': 7200, 'Entregado': 'Sí'},
             {'Ciudad': 'Iquique', 'Producto': 'Libros', 'Costo': 5300, 'Entregado': 'No'}]
df = pd.DataFrame(registros)
print(df)
```

```python
# ── Slide 9: Desde archivos externos ─────────────────────────────────────────

df = pd.read_csv('pedidos_logistica.csv')
```

```python
# ── Slide 9: Personalizando el índice ────────────────────────────────────────

data = {'Ciudad': ['Santiago', 'Valparaíso', 'Concepción'],
        'Producto': ['Libros', 'Ropa', 'Calzado'],
        'Costo': [4500, 6700, 8900],
        'Entregado': ['Sí', 'No', 'Sí']}
df = pd.DataFrame(data, index=['Pedido1', 'Pedido2', 'Pedido3'])
print(df)
```

---
## Slide 10 – Selección de Columnas

```python
# ── Slide 10: Acceso por nombre (forma preferida) ────────────────────────────

df['Ciudad']
```

```python
# ── Slide 10: Notación punto (no recomendada para nombres con espacios) ───────

df.Ciudad
```

```python
# ── Slide 10: Para seleccionar múltiples columnas ────────────────────────────

df[['Ciudad', 'Costo']]
```

---
## Slide 11 – Selección de Filas

```python
# ── Slide 11: Por posición (iloc) ────────────────────────────────────────────

df.iloc[0]   # Primera fila
```

```python
# ── Slide 11: Por etiqueta del índice (loc) ───────────────────────────────────

df.loc['Pedido2']
```

```python
# ── Slide 11: Varias filas por rango ─────────────────────────────────────────

df[1:3]
```

---
## Slide 12 – Selección de Elementos Específicos (loc / iloc)

```python
# ── Slide 12: Acceso por etiqueta (loc) ──────────────────────────────────────

df.loc[['Pedido1', 'Pedido3'], ['Ciudad', 'Costo']]
```

```python
# ── Slide 12: Acceso por posición (iloc) ─────────────────────────────────────

df.iloc[0, 2]   # primera fila, tercera columna
```

```python
# ── Slide 12: Celda específica por etiqueta ───────────────────────────────────

df.loc['Pedido1', 'Costo']
```

---
## Slide 14 – Selección Condicional en un DataFrame

```python
# ── Slide 14: Ejemplo — pedidos con costo mayor a $6000 ──────────────────────

df[df['Costo'] > 6000]
```

---
## Slide 15 – Filtrado por Múltiples Condiciones

```python
# ── Slide 15: Operador & ("y" lógico) ────────────────────────────────────────
# Cada condición debe ir entre paréntesis

df[(df['Costo'] > 6000) & (df['Entregado'] == 'No')]
```

## Slides 16–20 – Ejercicio Guiado: Empresa de Distribución Electrónica

**Objetivo (Slide 17):** Identificar pedidos no entregados con valor superior a $50.000.

```python
# ── Slide 18: Crear el DataFrame base ────────────────────────────────────────

import pandas as pd
data = {'ID Pedido': [1001, 1002, 1003, 1004, 1005],
        'Cliente': ['Juan', 'Sofía', 'Camila', 'Ignacio', 'Fernanda'],
        'Valor ($)': [62000, 48000, 71000, 32000, 90000],
        'Entregado': ['No', 'Sí', 'No', 'Sí', 'No']}
df = pd.DataFrame(data)
df.index = ['P1', 'P2', 'P3', 'P4', 'P5']   # Etiquetas personalizadas para filas
print(df)
```

```python
# ── Slide 19: Visualizar la tabla completa ────────────────────────────────────

print(df)
```

```python
# ── Slide 19: Seleccionar solo la columna Cliente ─────────────────────────────

df['Cliente']
```

```python
# ── Slide 19: Mostrar la fila con índice 'P3' ─────────────────────────────────

df.loc['P3']
```

```python
# ── Slide 19: Obtener el valor del pedido P5 en la columna Valor ($) ──────────

df.loc['P5', 'Valor ($)']
```

```python
# ── Slide 20: Filtrar pedidos con valor > $50.000 y no entregados ─────────────

filtro = (df['Valor ($)'] > 50000) & (df['Entregado'] == 'No')
df[filtro]
```

## Ejercicio Guiado con datos reales — `pedidos_logistica.csv`

Aplicamos las mismas operaciones al dataset real del desafío inicial.

```python
# ── Guiado con datos reales, Paso 1: Cargar y explorar ───────────────────────

df_log = pd.read_csv('pedidos_logistica.csv')
print(df_log.head())
print(f'\nDimensiones: {df_log.shape}')
```

```python
# ── Guiado, Paso 2: Selección de columnas (Slide 10) ─────────────────────────

# Una columna
print(df_log['ciudad'].head())
print()

# Múltiples columnas
print(df_log[['ciudad', 'costo', 'entregado']].head())
```

```python
# ── Guiado, Paso 3: Selección de filas y celdas (Slides 11–12) ───────────────

# Primera fila (posicional)
print('Primera fila:')
print(df_log.iloc[0])
print()

# Filas 5 a 9 (rango)
print('Filas 5 a 9:')
print(df_log.iloc[5:10])
print()

# Celda específica: fila 2, columna 'costo'
print('Costo del pedido en posición 2:', df_log.iloc[2, df_log.columns.get_loc('costo')])
```

```python
# ── Guiado, Paso 4: Filtrado condicional (Slides 14–15) ──────────────────────
# Misión: envíos costosos (> $50.000) que aún no se han entregado

filtro_critico = (df_log['costo'] > 50000) & (df_log['entregado'] == 'No')
df_criticos    = df_log[filtro_critico]

print(f'Total de pedidos:                          {len(df_log)}')
print(f'Pedidos costosos sin entregar (> $50.000): {len(df_criticos)}')
print()
print(df_criticos[['id_pedido','ciudad','producto','costo','entregado']].to_string(index=False))
```
