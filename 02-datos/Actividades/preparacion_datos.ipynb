```python
import pandas as pd

# DataFrame con información de empleados
data_empleados = {
    'id_empleado': [101, 102, 103, 104, 105, 106, 107, 108],
    'nombre': ['Ana', 'Luis', 'Sofía', 'Carlos', 'Elena', 'Pedro', 'Laura', 'David'],
    'salario': [55000, 58000, 72000, 48000, 75000, 85000, 62000, 88000],
    'id_departamento': ['V_01', 'V_01', 'T_01', 'M_01', 'T_01', 'D_01', 'M_01', 'D_01']
}

df_empleados = pd.DataFrame(data_empleados)

df_empleados.head()
```

```python
# DataFrame con información de departamentos
data_departamentos = {
    'id_departamento': ['V_01', 'T_01', 'M_01', 'D_01'],
    'nombre_depto': ['Ventas', 'Tecnología', 'Marketing', 'Dirección'],
    'ubicacion': ['Norte', 'Sur', 'Norte', 'Norte']
}

df_departamentos = pd.DataFrame(data_departamentos)

df_departamentos.head()
```

```python
# Unión de DataFrames

df_completo = pd.merge(
    df_empleados,
    df_departamentos,
    on='id_departamento',
    how='left'
)

print("DataFrame completo:")
print(df_completo)
```

```python
# Agrupamiento y promedio

salario_promedio = (
    df_completo
    .groupby(['nombre_depto', 'ubicacion'])['salario']
    .mean()
    .reset_index()
)

print("\nSalario promedio por departamento y ubicación:")
print(salario_promedio)
```

```python
# Tabla pivote

tabla_pivote = salario_promedio.pivot(
    index='nombre_depto',
    columns='ubicacion',
    values='salario'
)

print("\nTabla pivote final:")
print(tabla_pivote)
```
