```python
import pandas as pd
import numpy as np

df = pd.read_csv('datos_climaticos.csv')

df.head()
```

```python
datos_numericos = df[['temperatura_celsius', 'humedad_relativa']].to_numpy()

primera_hora = datos_numericos[:12]
print("\n--- a. Primera hora de mediciones (12 filas) ---")
print(primera_hora)
```

```python
temperaturas = datos_numericos[:, 0] 
temps_mayores_25 = temperaturas[temperaturas > 25.0]
print("\n--- b. Temperaturas mayores a 25.0°C ---")
print(temps_mayores_25)
```

```python
temperatura_fahrenheit = (temperaturas * (9/5)) + 32
print("\n--- c. Temperaturas convertidas a Fahrenheit ---")
print(temperatura_fahrenheit)
```

```python
primera_medicion = datos_numericos[0]   # Primera fila
ultima_medicion = datos_numericos[-1]   # Última fila
matriz_2x2 = np.array([primera_medicion, ultima_medicion])
print("\n--- d. Matriz 2x2 (Primera y última medición de temp y humedad) ---")
print(matriz_2x2)
```

```python
np.savetxt('temperaturas_fahrenheit.csv', temperatura_fahrenheit, delimiter=',', header='temperatura_fahrenheit', comments='')

print("\n¡Proceso finalizado! El archivo 'temperaturas_fahrenheit.csv' se guardó con éxito.")
```
