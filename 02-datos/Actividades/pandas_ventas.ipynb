```python
import pandas as pd

datos_ventas = {
    'Producto': [
        'Laptop', 'Mouse', 'Teclado',
        'Monitor', 'Webcam', 'Audífonos',
        'Laptop', 'Mouse', 'Mouse'
    ],
    'Categoría': [
        'Electrónica', 'Accesorios', 'Accesorios',
        'Electrónica', 'Accesorios', 'Accesorios',
        'Electrónica', 'Accesorios', 'Accesorios'
    ],
    'Precio': [1200, 25, 45, 300, 50, 75, 1250, 27, 26],
    'Unidades_Vendidas': [5, 30, 25, 10, 20, 15, 3, 40, 35]
}

df_ventas = pd.DataFrame(datos_ventas)

df_ventas.head()
```

```python
# 1. EXPLORACIÓN INICIAL

print("=== Información del DataFrame ===")
print(df_ventas.info())

print("\n=== Primeras 5 filas ===")
print(df_ventas.head())

print("\n=== Resumen estadístico ===")
print(df_ventas.describe())
```

```python
# 2. SELECCIÓN Y FILTRADO

print("\n=== Columna Producto (Serie) ===")
print(df_ventas["Producto"])

print("\n=== Columnas Producto y Precio ===")
print(df_ventas[["Producto", "Precio"]])

print("\n=== Ventas con Precio > 100 ===")
ventas_mayor_100 = df_ventas[df_ventas["Precio"] > 100]
print(ventas_mayor_100)
```

```python
# 3. SUMARIZACIÓN DE DATOS

print("\n=== Precio promedio ===")
precio_promedio = df_ventas["Precio"].mean()
print(precio_promedio)

print("\n=== Total productos vendidos categoría Accesorios ===")
total_accesorios = (
    df_ventas[df_ventas["Categoría"] == "Accesorios"]
    ["Unidades_Vendidas"]
    .sum()
)
print(total_accesorios)

print("\n=== Cantidad de veces que aparece cada producto ===")
conteo_productos = df_ventas["Producto"].value_counts()
print(conteo_productos)
```
