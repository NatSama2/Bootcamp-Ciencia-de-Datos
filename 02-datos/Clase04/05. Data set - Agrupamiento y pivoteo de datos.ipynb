## Slide 7 – Indexación Jerárquica con `groupby`

```python
# ── Slide 7: Ejemplo de indexación jerárquica ─────────────────────────────────

import pandas as pd
# Datos ficticios de ventas
data = {'Ciudad': ['Santiago', 'Santiago', 'Valparaíso', 'Valparaíso', 'Temuco'],
        'Semana': [1, 2, 1, 2, 1],
        'Ventas': [10000, 15000, 8000, 12000, 5000]}
df = pd.DataFrame(data)
df_grouped = df.groupby(['Ciudad', 'Semana']).sum()
print(df_grouped)
```

---
## Slide 10 – Agrupamiento con `groupby`

```python
# ── Slide 10: Ejemplo básico — groupby con sum() ──────────────────────────────

df.groupby('Ciudad')['Ventas'].sum()
```

```python
# ── Slide 10: Ejemplo avanzado — groupby con agg() múltiple ──────────────────
# Genera una tabla agregada con múltiples estadísticas por grupo

df.groupby(['Ciudad', 'Semana'])['Ventas'].agg(['sum', 'mean', 'count'])
```

---
## Slide 11 – `as_index=False` y `reset_index()`

```python
# ── Slide 11: as_index=False — mantiene formato de tabla plana ────────────────
# Útil para visualización o exportación directa

df.groupby('Ciudad', as_index=False)['Ventas'].sum()
```

```python
# ── Slide 11: reset_index() — aplana el resultado del groupby ────────────────
# Necesario cuando se usa agg() y se quiere exportar

df.groupby(['Ciudad', 'Semana'])['Ventas'].agg(['sum', 'mean', 'count']).reset_index()
```

## Slides 12–15 – Actividad Guiada: Agrupando para Decidir

**Objetivo (Slide 13):** Total de ventas por semana y ciudad, promedio semanal y número de transacciones.

**Archivo:** `ventas_semanales.csv`

```python
# ── Slide 14, Paso 1: Cargar el archivo CSV ───────────────────────────────────

import pandas as pd
df = pd.read_csv('ventas_semanales.csv')
```

```python
# ── Slide 14, Paso 2: Agrupar por ciudad y semana ────────────────────────────

resumen = df.groupby(['Ciudad', 'Semana'])['Ventas'].agg(['sum', 'mean', 'count'])
print(resumen)
```

```python
# ── Slide 14, Paso 3: Convertir en DataFrame plano y exportar ─────────────────

resumen.reset_index().to_csv('resumen_agrupado.csv', index=False)
print('✅ resumen_agrupado.csv exportado')
```

```python
# ── Slide 15: Preguntas de reflexión — respuestas con código ─────────────────

resumen_plano = resumen.reset_index()

# ¿Qué ciudad tuvo mayor total de ventas?
total_ciudad = df.groupby('Ciudad')['Ventas'].sum().sort_values(ascending=False)
print('Total de ventas por ciudad:')
print(total_ciudad)

# ¿Cuál semana tuvo más transacciones en total?
trans_semana = df.groupby('Semana')['Ventas'].count().sort_values(ascending=False)
print('\nTransacciones por semana:')
print(trans_semana)

# ¿Cuál producto tuvo mayor venta promedio?
prom_producto = df.groupby('Producto')['Ventas'].mean().sort_values(ascending=False)
print('\nVenta promedio por producto:')
print(prom_producto.round(0))
```

## Slides 16–20 – Actividad Autónoma: Resumen de Atención Médica

**Contexto (Slide 17):** Resumen de atenciones médicas por centro y tipo de consulta.

---
**Completa las celdas marcadas con `# 📝 TU CÓDIGO AQUÍ`**

```python
# ── Slide 18: Crear el DataFrame base ────────────────────────────────────────

import pandas as pd
datos = {'Centro': ['Clínica Norte', 'Clínica Norte', 'Clínica Sur', 'Clínica Sur',
                    'Clínica Centro', 'Clínica Centro', 'Clínica Norte', 'Clínica Sur'],
         'TipoConsulta': ['General', 'Especialista', 'General', 'Especialista',
                          'General', 'Especialista', 'General', 'Especialista'],
         'Atenciones': [120, 80, 100, 90, 110, 70, 130, 85]}
df = pd.DataFrame(datos)
print(df)
```

```python
# ── Slide 19: Agrupar por 'Centro' y 'TipoConsulta' con agg() ────────────────

resumen = df.groupby(['Centro', 'TipoConsulta'])['Atenciones'].agg(['sum', 'mean', 'count'])
print(resumen)
```

```python
# ── Slide 20: Exportar el resultado como resumen_atenciones.csv ───────────────

resumen.reset_index().to_csv('resumen_atenciones.csv', index=False)
print('✅ resumen_atenciones.csv exportado')
```
