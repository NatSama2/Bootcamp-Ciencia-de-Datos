## Slides 5–7 – Variables Cualitativas

Categóricas: describen atributos, no magnitudes.
- **Nominales:** sin orden (color, género, región)
- **Ordinales:** con orden (nivel educativo: básico → medio → superior)

```python
# ── Slide 6: Representación en pandas — dtype object ─────────────────────────

df['Modalidad'] = pd.Series(['Online', 'Presencial', 'Online'])
print(df['Modalidad'].dtype)
```

```python
# ── Slide 6: Convertir a tipo category (ahorra memoria, acelera procesos) ─────

df['Modalidad'] = df['Modalidad'].astype('category')
```

```python
# ── Slide 7: Buenas prácticas — value_counts() para validar categorías ────────
# Detecta variaciones como 'RM', 'Región Metropolitana', 'rm'

df_real = pd.read_csv('rendimiento_academico.csv')
print('Distribución de regiones (con posibles errores):')
print(df_real['region'].value_counts())
```

## Slides 10–11 – Variables Cuantitativas

Expresan magnitudes. En pandas: `int64`, `float64`.
- **Continua:** cualquier valor en un rango (peso, temperatura)
- **Discreta:** solo enteros (número de hijos, cantidad de intentos)

```python
# ── Slide 10: Tipo de dato de una variable cuantitativa ───────────────────────

df['Edad'] = pd.Series([24, 35, 41])
print(df['Edad'].dtype)
```

```python
# ── Slide 11: Exploración básica de variables cuantitativas ───────────────────

df_real = pd.read_csv('rendimiento_academico.csv')
df_real.describe()          # Estadísticas descriptivas
df_real['nota'].hist()      # Histograma (si se usa con matplotlib)
plt.xlabel('Nota')
plt.title('Histograma de notas (incluye outliers)')
plt.show()
```

---
## Slide 14 – Detección de Outliers: Método IQR

```python
# ── Slide 14: Método de los cuartiles (IQR) ───────────────────────────────────

Q1 = df['Nota'].quantile(0.25)
Q3 = df['Nota'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers = df[(df['Nota'] < limite_inferior) | (df['Nota'] > limite_superior)]
```

---
## Slide 15 – Más Métodos: Z-score y Boxplot

```python
# ── Slide 15: Z-score (requiere distribución normal) ─────────────────────────

from scipy import stats
import numpy as np
z_scores = np.abs(stats.zscore(df['Edad']))
df[z_scores > 3]   # Valores cuyo z-score supera 3
```

```python
# ── Slide 15: Visualización — boxplot ────────────────────────────────────────

import matplotlib.pyplot as plt
df.boxplot(column='Edad')
plt.show()
```

---
## Slide 16 – Filtrado de Outliers

```python
# ── Slide 16: Filtrar conservando solo registros dentro de los límites IQR ────

df_filtrado = df[(df['Nota'] >= limite_inferior) & (df['Nota'] <= limite_superior)]
```

## Slides 17–21 – Actividad Guiada: Rendimiento Académico

**Objetivo (Slide 18):** Clasificar variables, detectar y filtrar outliers en notas.

```python
# ── Slide 19: Crear el DataFrame base ────────────────────────────────────────

import pandas as pd
df = pd.DataFrame({
    'Estudiante': ['Pedro', 'Laura', 'Ignacio', 'Fernanda', 'Javiera', 'Luis', 'Camila'],
    'Nota': [5.5, 6.2, 10.8, 4.9, 3.7, -2.0, 5.8],
    'Modalidad': ['Presencial', 'Online', 'Online', 'Presencial', 'Online', 'Online', 'Presencial']
})
```

```python
# ── Slide 20: Identificar el tipo de cada variable ────────────────────────────

print(df.dtypes)
```

```python
# ── Slide 20: Detectar outliers con el método de los cuartiles ───────────────

Q1 = df['Nota'].quantile(0.25)
Q3 = df['Nota'].quantile(0.75)
IQR = Q3 - Q1
limite_inf = Q1 - 1.5 * IQR
limite_sup = Q3 + 1.5 * IQR
outliers = df[(df['Nota'] < limite_inf) | (df['Nota'] > limite_sup)]
print(outliers)
```

```python
# ── Slide 21: Filtrar los datos válidos ──────────────────────────────────────

df_limpio = df[(df['Nota'] >= limite_inf) & (df['Nota'] <= limite_sup)]
```

```python
# ── Slide 21: Comparar antes y después ───────────────────────────────────────

print('=== ANTES del filtrado ===')
print(df[['Estudiante','Nota']])
print(f'Promedio con outliers: {df["Nota"].mean():.2f}')
print()
print('=== DESPUÉS del filtrado ===')
print(df_limpio[['Estudiante','Nota']])
print(f'Promedio sin outliers: {df_limpio["Nota"].mean():.2f}')
print(f'\nRegistros eliminados: {len(df) - len(df_limpio)}')
```

## Actividad Guiada ampliada — `rendimiento_academico.csv` (n=300)

Aplicamos clasificación de variables, IQR, z-score y boxplot al dataset real del desafío.

```python
# Clasificación de variables del dataset real
df_real = pd.read_csv('rendimiento_academico.csv')

print('=== Tipos de datos ===')
print(df_real.dtypes)
print()
print('Variables cualitativas (object):',
      df_real.select_dtypes(include='object').columns.tolist())
print('Variables cuantitativas (numérico):',
      df_real.select_dtypes(include='number').columns.tolist())
```

```python
# Detección de outliers en 'nota' con IQR
Q1  = df_real['nota'].quantile(0.25)
Q3  = df_real['nota'].quantile(0.75)
IQR = Q3 - Q1
lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR

print(f'Q1={Q1:.2f}  Q3={Q3:.2f}  IQR={IQR:.2f}')
print(f'Límite inferior: {lim_inf:.2f}  |  Límite superior: {lim_sup:.2f}')

outliers_iqr = df_real[(df_real['nota'] < lim_inf) | (df_real['nota'] > lim_sup)]
print(f'\nOutliers detectados por IQR: {len(outliers_iqr)}')
print(outliers_iqr[['id_estudiante','nota']])
```

```python
# Detección con Z-score (columna horas_estudio, distribución más normal)
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df_real['nota'].dropna()))
outliers_z = df_real.loc[df_real['nota'].dropna().index[z_scores > 3]]
print(f'Outliers detectados por Z-score (|z|>3): {len(outliers_z)}')
print(outliers_z[['id_estudiante','nota']])
```

```python
# Boxplot para visualización
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Con outliers
axes[0].boxplot(df_real['nota'].dropna())
axes[0].set_title('Notas — con outliers', fontsize=12)
axes[0].set_ylabel('Nota')
axes[0].axhline(7.0, color='red', linestyle='--', linewidth=1.5, label='Límite superior escala (7.0)')
axes[0].axhline(1.0, color='orange', linestyle='--', linewidth=1.5, label='Límite inferior escala (1.0)')
axes[0].legend(fontsize=8)

# Sin outliers
df_limpio = df_real[(df_real['nota'] >= lim_inf) & (df_real['nota'] <= lim_sup)]
axes[1].boxplot(df_limpio['nota'])
axes[1].set_title(f'Notas — sin outliers (n={len(df_limpio)})', fontsize=12)
axes[1].set_ylabel('Nota')

plt.suptitle('Impacto del filtrado de outliers en la distribución de notas', fontsize=13)
plt.tight_layout()
plt.show()

print(f'Promedio CON outliers:  {df_real["nota"].mean():.4f}')
print(f'Promedio SIN outliers:  {df_limpio["nota"].mean():.4f}')
print(f'Diferencia:             {abs(df_real["nota"].mean() - df_limpio["nota"].mean()):.4f}')
```

```python
# Unificar variaciones de región (Slide 7: buenas prácticas)
mapa_regiones = {
    'Región Metropolitana': 'RM', 'rm': 'RM', 'r.m.': 'RM',
    'V Región': 'Valparaíso', 'Valparaiso': 'Valparaíso',
    'Biobio': 'Bio-Bío', 'Bio Bío': 'Bio-Bío', 'bio-bio': 'Bio-Bío',
    'La Araucania': 'Araucanía',
    'Maule ': 'Maule', 'los lagos': 'Los Lagos',
}
df_limpio['region'] = df_limpio['region'].replace(mapa_regiones)
print(f'Regiones únicas después de unificar: {df_limpio["region"].nunique()}')
print(df_limpio['region'].value_counts())
```
