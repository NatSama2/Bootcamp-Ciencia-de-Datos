## Slide 6 – Sintaxis de `pivot()`

- `index`: columna que permanecerá como índice de fila
- `columns`: valores únicos que se convertirán en columnas
- `values`: valores que se ubicarán en la intersección fila/columna

> ⚠️ Solo funciona si **no hay duplicados** en la combinación index + columns.

```python
# ── Slide 6: Sintaxis básica de pivot() ──────────────────────────────────────

DataFrame.pivot(index='...', columns='...', values='...')
```

---
## Slide 7 – Ejemplo Práctico de `pivot()`

```python
# ── Slide 7: Ejemplo práctico de pivot() ─────────────────────────────────────

import pandas as pd
data = {'curso': ['Python', 'Python', 'SQL', 'SQL'],
        'mes': ['Enero', 'Febrero', 'Enero', 'Febrero'],
        'participaciones': [25, 30, 18, 22]}
df = pd.DataFrame(data)
pivotado = df.pivot(index='curso', columns='mes', values='participaciones')
print(pivotado)
```

---
## Slide 8 – Limitación de `pivot()`: Duplicados

```python
# ── Slide 8: Error con duplicados en pivot() ──────────────────────────────────
# ValueError: Index contains duplicate entries, cannot reshape
# Solución: usar pivot_table()

data_dup = {'curso': ['Python', 'Python', 'SQL', 'SQL', 'Python'],
            'mes':   ['Enero',  'Enero',  'Enero', 'Febrero', 'Febrero'],
            'participaciones': [25, 10, 18, 22, 30]}
df_dup = pd.DataFrame(data_dup)

try:
    df_dup.pivot(index='curso', columns='mes', values='participaciones')
except ValueError as e:
    print(f'❌ Error: {e}')
    print('→ Solución: usar pivot_table()')
```

---
## Slide 9 – Sintaxis de `pivot_table()`

```python
# ── Slide 9: Sintaxis extendida de pivot_table() ──────────────────────────────

DataFrame.pivot_table(index='...', columns='...', values='...', aggfunc='...', fill_value=0)
```

---
## Slide 10 – Ejemplo de `pivot_table()`

```python
# ── Slide 10: Ejemplo completo de pivot_table() ───────────────────────────────

df = pd.DataFrame({
    'curso': ['Python', 'Python', 'SQL', 'SQL', 'Python'],
    'mes': ['Enero', 'Febrero', 'Enero', 'Febrero', 'Enero'],
    'asistentes': [25, 30, 18, 22, 20]
})
pivotado = df.pivot_table(
    index='curso',
    columns='mes',
    values='asistentes',
    aggfunc='sum',
    fill_value=0
)
print(pivotado)
```

---
## Slide 13 – Sintaxis de `melt()` (Despivoteo)

```python
# ── Slide 13: Sintaxis de melt() ─────────────────────────────────────────────
# id_vars: columna(s) que permanecerán fijas
# var_name: nombre para las columnas convertidas en fila
# value_name: nombre de la columna de valores numéricos

pd.melt(df, id_vars=[...], var_name='...', value_name='...')
```

---
## Slide 14 – Ejemplo de `melt()`

```python
# ── Slide 14: Ejemplo completo de melt() ─────────────────────────────────────

pivotado = pivotado.reset_index()
despivotado = pd.melt(
    pivotado,
    id_vars='curso',
    var_name='mes',
    value_name='asistentes'
)
print(despivotado)
```

## Slides 15–18 – Actividad Guiada: Asistencias a Talleres

**Objetivo (Slide 16):** Pivotear con `pivot_table()` y luego revertir con `melt()`.

**Archivo:** `asistencias_talleres.csv`

```python
# ── Slide 17: Cargar el archivo ───────────────────────────────────────────────

import pandas as pd
df = pd.read_csv('asistencias_talleres.csv')
```

```python
df
```

```python
# ── Slide 17: Pivotear los datos ──────────────────────────────────────────────

tabla = df.pivot_table(
    index='taller',
    columns='mes',
    values='asistentes',
    aggfunc='sum'
)
print('Tabla pivoteada (formato ancho):')
print(tabla)
```

```python
# ── Slide 18: Despivotear los datos ──────────────────────────────────────────

tabla = tabla.reset_index()
df_original = pd.melt(tabla,
                      id_vars='taller',
                      var_name='mes',
                      value_name='asistentes')
print('DataFrame revertido (formato largo):')
print(df_original.head(12))
```

```python
# ── Slide 18: Guardar resultados ──────────────────────────────────────────────

tabla.to_csv('asistencias_pivot.csv', index=False)
df_original.to_csv('asistencias_revertidas.csv', index=False)
print('✅ asistencias_pivot.csv y asistencias_revertidas.csv exportados')
```

```python
# ── Slide 19: Verificación con margins=True (totales por fila y columna) ──────

df = pd.read_csv('asistencias_talleres.csv')
tabla_con_totales = df.pivot_table(
    index='taller',
    columns='mes',
    values='asistentes',
    aggfunc='sum',
    fill_value=0,
    margins=True,
    margins_name='TOTAL'
)
print(tabla_con_totales)
```

```python
import pandas as pd

print("=== 1. Cargar archivo ===")

df = pd.read_csv('ventas_mensuales.csv')

print("Archivo cargado correctamente")
print(f"Cantidad de filas: {len(df)}")


print("\n=== 2. Revisar columnas ===")

print("Columnas originales:")
print(df.columns.tolist())

# Limpiar espacios en nombres de columnas
df.columns = df.columns.str.strip()

print("\nColumnas después de limpiar espacios:")
print(df.columns.tolist())


print("\n=== 3. Mostrar primeras filas ===")
print(df.head())


print("\n=== 4. Verificar columnas necesarias ===")

if 'mes' not in df.columns:
    print("❌ No existe la columna 'mes'")
    
if 'sucursal' not in df.columns:
    print("❌ No existe la columna 'sucursal'")
    
if 'ventas' not in df.columns:
    print("❌ No existe la columna 'ventas'")


# Continuar solo si existen
if all(col in df.columns for col in ['mes', 'sucursal', 'ventas']):

    print("\n=== 5. Aplicar orden de meses ===")

    orden_meses = [
        'Enero', 'Febrero', 'Marzo', 'Abril',
        'Mayo', 'Junio', 'Julio', 'Agosto',
        'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
    ]

    # Limpiar espacios dentro de los valores
    df['mes'] = df['mes'].astype(str).str.strip()

    print("Meses encontrados:")
    print(df['mes'].unique())

    df['mes'] = pd.Categorical(
        df['mes'],
        categories=orden_meses,
        ordered=True
    )

    print("\nMes convertido a categórico")
    print(df['mes'].head())


    print("\n=== 6. Crear tabla pivot ===")

    resumen = df.pivot_table(
        index='sucursal',
        columns='mes',
        values='ventas',
        aggfunc='sum'
    )

    print("Pivot generado:")
    print(resumen)


    print("\n=== 7. Resetear índice ===")

    resumen = resumen.reset_index()

    print(resumen.head())


    print("\n=== 8. Despivotar con melt ===")

    df_largo = pd.melt(
        resumen,
        id_vars='sucursal',
        var_name='mes',
        value_name='ventas'
    )

    print("Resultado despivotado:")
    print(df_largo.head(10))


    print("\n=== 9. Guardar archivos ===")

    resumen.to_csv('pivotado.csv', index=False)
    df_largo.to_csv('despivotado.csv', index=False)

    print("✓ Archivo pivotado.csv guardado")
    print("✓ Archivo despivotado.csv guardado")


    print("\n=== 10. Resumen final ===")

    print(f"Tamaño pivotado: {resumen.shape}")
    print(f"Tamaño despivotado: {df_largo.shape}")

else:
    print("\n⚠️ Revisa los nombres de columnas del CSV")
```
