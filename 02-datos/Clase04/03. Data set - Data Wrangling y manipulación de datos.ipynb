## Slides 5–6 – ¿Qué es Data Wrangling?

Proceso de preparar datos crudos para análisis: limpiar, reorganizar, convertir formatos, corregir inconsistencias.

> Esta etapa consume entre el **60% y el 80%** del tiempo de un proyecto de ciencia de datos *(Lohr, 2014)*.

---
## Slides 7–9 – Principales Tareas del Data Wrangling

| Tarea | Descripción |
|---|---|
| Evaluación de calidad | Diagnóstico de datos ausentes, formatos inconsistentes |
| Limpieza | Imputación, filtrado de outliers, eliminación de duplicados |
| Transformación | Conversión de tipos, renombrar columnas, normalización |
| Enriquecimiento | Agregación, combinación con otras fuentes, nuevas variables |
| Reformateo estructural | Reordenar, pivotear, cambiar índices |
| Reducción / muestreo | Filtrado, segmentación, muestras representativas |

## Slide 11 – Ordenamiento con `sort_values()` y `sort_index()`

```python
# ── Slide 11: sort_values() — ordenar por valores de una columna ──────────────
# by: columna de referencia | ascending: True=ascendente, False=descendente

df.sort_values(by='precio', ascending=False)
```

```python
# ── Slide 11: sort_index() — ordenar por índice ───────────────────────────────
# Útil con índices jerárquicos o antes de groupby

df.sort_index(ascending=True)
```

---
## Slide 12 – Reordenar Columnas

```python
# ── Slide 12: Reordenar columnas manualmente ──────────────────────────────────
# ¡Importante! Usar df.copy() para no modificar el DataFrame original

df = df[['fecha', 'producto', 'precio', 'cantidad']]
```

---
## Slide 13 – Muestreo Aleatorio: `sample(n)`

```python
# ── Slide 13: Muestreo simple aleatorio — n filas ─────────────────────────────

df.sample(n=10)
```

---
## Slide 14 – Muestreo con Porcentaje y Semilla de Reproducibilidad

```python
# ── Slide 14: Muestreo con porcentaje — frac ──────────────────────────────────
# Extrae el 20% del total de filas

df.sample(frac=0.2)
```

```python
# ── Slide 14: Fijar semilla para reproducibilidad — random_state ──────────────
# Garantiza que siempre se obtenga la misma muestra en cada ejecución

df.sample(n=5, random_state=42)
```

## Slides 15–17 – Actividad Guiada: Informe de Ventas

**Objetivo (Slide 16):** Ordenar, reordenar columnas y muestrear para preparar un informe ejecutivo.

**Archivo:** `ventas_sin_orden.csv`

```python
# ── Slide 17: Cargar el archivo ───────────────────────────────────────────────

import pandas as pd
df = pd.read_csv('ventas_sin_orden.csv')
print('Estado inicial:')
print(df.head())
print(f'\nColumnas actuales: {list(df.columns)}')
```

```python
# ── Slide 17: Ordenar por 'fecha' en orden ascendente ────────────────────────

df = df.sort_values(by='fecha', ascending=True)
print('Primeras filas después de ordenar por fecha:')
print(df.head())
```

```python
# ── Slide 17: Reordenar columnas ──────────────────────────────────────────────

df = df[['producto', 'cantidad', 'precio_unitario', 'fecha', 'vendedor']]
print('Columnas reordenadas:')
print(df.head())
```

```python
# ── Slide 17: Muestra aleatoria del 25% ──────────────────────────────────────

muestra_ventas = df.sample(frac=0.25, random_state=42)
print(f'Muestra generada: {len(muestra_ventas)} filas ({len(muestra_ventas)/len(df)*100:.0f}% del total)')
print(muestra_ventas.head())
```

```python
# ── Slide 17: Exportar la muestra como informe_semana.csv ────────────────────

muestra_ventas.to_csv('informe_semana.csv', index=False)
print('✅ informe_semana.csv exportado correctamente')
```
