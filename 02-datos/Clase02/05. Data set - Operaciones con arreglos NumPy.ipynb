## PARTE 1 — Indexación y selección de elementos

### 1.1 Indexación en arreglos unidimensionales

```python
velocidades = np.array([30, 45, 55, 60, 48, 52])
print('Velocidades:', velocidades)

# Acceso por posición (índice positivo)
print()
print(f'Posición 0 (primero): {velocidades[0]}')
print(f'Posición 3:           {velocidades[3]}')

# Índices negativos (desde el final)
print()
print(f'Posición -1 (último):           {velocidades[-1]}')
print(f'Posición -3 (tercero desde fin): {velocidades[-3]}')
```

### 1.2 Slicing (rebanado)

```python
print('Posiciones 1 a 3 [1:4]:   ', velocidades[1:4])   # [45 55 60]
print('Primeros tres [:3]:        ', velocidades[:3])    # [30 45 55]
print('Desde posición 3 [3:]:     ', velocidades[3:])   # [60 48 52]
print('Toda la secuencia [:]:     ', velocidades[:])    # completo
print('Paso de 2 [::2]:           ', velocidades[::2])  # [30 55 48]
```

### 1.3 Indexación en matrices bidimensionales

```python
matriz = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])
print('Matriz:')
print(matriz)

print()
print(f'Elemento [1,2] (fila 1, col 2): {matriz[1, 2]}')  # 60
print(f'Fila 0 completa:                {matriz[0]}')
print(f'Columna 1 completa [:, 1]:      {matriz[:, 1]}')
print()
print('Submatriz filas 0-1, cols 1-2:')
print(matriz[0:2, 1:3])
```

### 1.4 Modificación de valores e indexación avanzada

```python
v = velocidades.copy()   # trabajamos sobre una copia

# Modificar un elemento
v[2] = 50
print('Después de v[2] = 50:', v)

# Reemplazar un segmento
v[0:3] = [35, 46, 51]
print('Después de reemplazar [0:3]:', v)

# Indexación avanzada: seleccionar por lista de posiciones
posiciones = [0, 2, 4]
print(f'Valores en posiciones {posiciones}: {v[posiciones]}')
```

## PARTE 2 — Selección condicional (indexación booleana)

### 2.1 Condición sobre vector

```python
temperaturas = np.array([22.5, 25.0, 19.8, 27.3, 21.0])

# Generar máscara booleana
condicion = temperaturas > 23
print('Máscara (> 23):', condicion)

# Usar la máscara como índice
print('Valores que cumplen:', temperaturas[condicion])

# Expresión compacta
print('Compacto:', temperaturas[temperaturas > 23])
```

### 2.2 Otras condiciones y múltiples criterios

```python
print('Menores a 21:  ', temperaturas[temperaturas < 21])
print('Igual a 25.0:  ', temperaturas[temperaturas == 25.0])
print('Distinto 22.5: ', temperaturas[temperaturas != 22.5])

print()
# Múltiples condiciones — paréntesis OBLIGATORIOS
# & = AND  |  | = OR  |  ~ = NOT
entre_20_26 = (temperaturas > 20) & (temperaturas < 26)
print('Entre 20 y 26 (AND):', temperaturas[entre_20_26])

extremos = (temperaturas < 21) | (temperaturas > 26)
print('Menores 21 o mayores 26 (OR):', temperaturas[extremos])
```

### 2.3 Condiciones sobre matrices

```python
m = np.array([[10, 20, 30], [5, 25, 35], [15, 18, 40]])
print('Matriz:')
print(m)

print()
print('Elementos > 20:', m[m > 20])       # vector con los que cumplen

# Máscara booleana reutilizable
mascara_pares = m % 2 == 0
print('Máscara pares:')
print(mascara_pares)
print('Elementos pares:', m[mascara_pares])
```

### 2.4 Asignación condicional (limpieza de datos)

```python
t = temperaturas.copy()
print('Antes:', t)

# Reemplazar todos los valores menores a 22 con el umbral mínimo
t[t < 22] = 22
print('Después (imputación mínima 22°C):', t)
```

## PARTE 3 — Referencia vs. Copia

### 3.1 El problema del aliasing

```python
a = np.array([1, 2, 3])
b = a               # b NO es una copia — apunta al mismo objeto

b[0] = 100

print('a:', a)      # ⚠️ [100   2   3] — ¡a también cambió!
print('b:', b)      # [100   2   3]
print('¿Comparten memoria?', np.shares_memory(a, b))
```

### 3.2 Copia real con `.copy()`

```python
a = np.array([1, 2, 3])
c = a.copy()        # c es completamente independiente

c[0] = 999

print('a:', a)      # [1 2 3] — no cambia
print('c:', c)      # [999   2   3]
print('¿Comparten memoria?', np.shares_memory(a, c))
```

### 3.3 Vista (view) con slicing sobre matrices

```python
matriz = np.array([[1, 2], [3, 4]])

# Slicing crea una VISTA, no una copia
submatriz = matriz[:, 0]
submatriz[0] = 100

print('Matriz original (afectada):')
print(matriz)       # [[100, 2], [3, 4]]

print()
# Solución: usar .copy() explícitamente
matriz2 = np.array([[1, 2], [3, 4]])
sub_copia = matriz2[:, 0].copy()
sub_copia[0] = 200

print('Matriz2 (NO afectada con .copy()):')
print(matriz2)      # [[1, 2], [3, 4]]
```

## PARTE 4 — Operaciones entre arreglos

### 4.1 Operaciones elemento a elemento

```python
a = np.array([10, 20, 30])
b = np.array([1,  2,  3])

print(f'a + b = {a + b}')
print(f'a - b = {a - b}')
print(f'a * b = {a * b}')
print(f'a / b = {a / b}')
print(f'a ** 2 = {a ** 2}')
```

### 4.2 Broadcasting

```python
# Un solo valor se "expande" automáticamente
a = np.array([10, 20, 30])
b = np.array([1])
print('Broadcasting [1]:', a + b)    # [11 21 31]

# Entre matriz y vector (compatible)
m1 = np.array([[1, 2], [3, 4]])
fila = np.array([1, 2])
print()
print('Matriz + vector fila:')
print(m1 + fila)    # [[2,4],[4,6]]

# Error si dimensiones incompatibles
try:
    np.array([10, 20, 30]) + np.array([1, 2])
except ValueError as e:
    print()
    print(f'❌ Error de broadcasting: {e}')
```

### 4.3 Operaciones lógicas y comparaciones entre arreglos

```python
a = np.array([10, 20, 30])
b = np.array([10, 15, 30])

print('a == b:', a == b)    # [ True False  True]
print('a >  b:', a >  b)    # [False  True False]
print('a != b:', a != b)    # [False  True False]
```

### 4.4 Funciones de agregación

```python
valores = np.array([10, 20, 30, 40])

print(f'Suma:                {np.sum(valores)}')
print(f'Promedio:            {np.mean(valores)}')
print(f'Mínimo:              {np.min(valores)}')
print(f'Máximo:              {np.max(valores)}')
print(f'Desv. estándar:      {np.std(valores):.4f}')
print(f'Mediana:             {np.median(valores)}')

# Combinación de arreglos
a = np.array([10, 20, 30])
b = np.array([10, 15, 30])
print()
print('Promedio elemento a elemento (a+b)/2:', (a + b) / 2)

# Promoción de tipos
x_int   = np.array([1, 2, 3])
y_float = np.array([0.5, 1.5, 2.5])
print('int + float:', x_int + y_float, '→ dtype:', (x_int + y_float).dtype)
```

## PARTE 5 — Operaciones con escalares

### 5.1 Operaciones básicas

```python
datos = np.array([10, 20, 30])

print('datos + 5  =', datos + 5)    # suma
print('datos - 3  =', datos - 3)    # resta
print('datos * 2  =', datos * 2)    # multiplicación
print('datos / 10 =', datos / 10)   # división (→ float)
print('datos ** 2 =', datos ** 2)   # potencia
print('datos % 7  =', datos % 7)    # módulo
print('datos // 3 =', datos // 3)   # división entera (trunca decimales)
```

### 5.2 Aplicaciones frecuentes

```python
# Ajuste de precios (+5%)
precios = np.array([100, 200, 300])
nuevos_precios = precios * 1.05
print('Precios con +5%:', nuevos_precios)

# Conversión de temperatura °C → °F
celsius    = np.array([0, 10, 20, 30, 40])
fahrenheit = celsius * 9/5 + 32
print('Celsius:   ', celsius)
print('Fahrenheit:', fahrenheit)

# Normalización min-max → escala 0 a 1
mediciones = np.array([5, 10, 15, 20, 25])
normalizados = (mediciones - np.min(mediciones)) / (np.max(mediciones) - np.min(mediciones))
print()
print('Normalizados [0,1]:', normalizados)

# Cuidado: booleanos como enteros
flags = np.array([True, False, True])
print()
print('Booleanos * 10:', flags * 10)   # [10 0 10]
```

## PARTE 6 — Funciones universales (ufuncs)

### 6.1 Funciones matemáticas elementales

```python
datos = np.array([1, 4, 9, 16])

print('Raíz cuadrada np.sqrt:', np.sqrt(datos))
print('Logaritmo natural  np.log:', np.round(np.log(datos), 4))
print('Exponencial  np.exp:', np.round(np.exp(datos[:3]), 2))   # limitado por espacio
print('Valor absoluto np.abs:', np.abs(np.array([-3, -1, 0, 2, 5])))

# Composición de funciones
resultado = np.log(np.sqrt(datos))
print()
print('log(sqrt(datos)):', np.round(resultado, 4))
```

### 6.2 Funciones estadísticas por eje en matrices

```python
matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print('Matriz:')
print(matriz)

print()
print('Suma total (sin eje):', np.sum(matriz))
print('Suma por columna (axis=0):', np.sum(matriz, axis=0))    # [5 7 9]
print('Suma por fila    (axis=1):', np.sum(matriz, axis=1))    # [6 15]
print()
print('Media por columna (axis=0):', np.mean(matriz, axis=0))
print('Media por fila    (axis=1):', np.mean(matriz, axis=1))
print('Máximo por columna (axis=0):', np.max(matriz, axis=0))
```

### 6.3 Funciones booleanas: `any`, `all`, `sum`

```python
datos = np.array([3, 7, 2, 9])

print('datos > 5:', datos > 5)
print('np.any(datos > 5):', np.any(datos > 5))   # ¿alguno cumple?
print('np.all(datos > 1):', np.all(datos > 1))   # ¿todos cumplen?
print('np.sum(datos > 5):', np.sum(datos > 5))   # ¿cuántos cumplen?
```

### 6.4 ufuncs directas: `np.add`, `np.multiply`

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print('np.add(x,y):       ', np.add(x, y))       # equivalente a x+y
print('np.multiply(x,y):  ', np.multiply(x, y))  # equivalente a x*y
print('np.power(x,2):     ', np.power(x, 2))     # equivalente a x**2
print('np.maximum(x,y):   ', np.maximum(x, y))   # máximo elemento a elemento
```

## PARTE 7 — Ejercicio guiado: Análisis de lecturas meteorológicas

**Situación:** Tu jefatura te entrega datos de 8 horas de una estación meteorológica (temperatura °C, humedad %, velocidad del viento km/h) y te pide analizar anomalías y estadísticas.

```python
# Matriz: 8 horas × 3 variables [temperatura, humedad, viento]
np.random.seed(42)
estacion = np.array([
    [18.5, 72, 15],
    [19.2, 68, 18],
    [21.0, 65, 22],
    [23.5, 60, 30],
    [24.8, 58, 35],
    [22.1, 63, 28],
    [20.3, 70, 20],
    [17.9, 75, 12]
])

print('Datos de la estación (8h × 3 variables):')
print(f'{"Hora":>6} {"Temp(°C)":>10} {"Humedad(%)": >12} {"Viento(km/h)":>14}')
print('-' * 48)
for i, fila in enumerate(estacion):
    print(f'{i:>6} {fila[0]:>10.1f} {fila[1]:>12.0f} {fila[2]:>14.0f}')
```

```python
# Paso 1: Estadísticas por variable (columna)
print('=== Estadísticas por variable ===')
variables = ['Temperatura', 'Humedad', 'Viento']
for i, var in enumerate(variables):
    col = estacion[:, i]
    print(f'{var:<14} → media: {np.mean(col):.1f}  min: {np.min(col):.1f}  max: {np.max(col):.1f}  std: {np.std(col):.2f}')
```

```python
# Paso 2: Extraer subconjuntos con condiciones

# Horas con temperatura mayor a 22°C
temp = estacion[:, 0]
horas_calor = np.where(temp > 22)[0]
print('Horas con temperatura > 22°C:', horas_calor)
print('Temperaturas:', temp[horas_calor])

print()
# Horas con viento fuerte (> 28 km/h)
viento = estacion[:, 2]
horas_viento = np.where(viento > 28)[0]
print('Horas con viento > 28 km/h:', horas_viento)
print('Viento:', viento[horas_viento])
```

```python
# Paso 3: Vector de umbrales de alerta
umbrales = np.array([23.0, 65.0, 30.0])   # [temp_max, humedad_min, viento_max]
print('Umbrales de alerta:', umbrales)

# Verificar si alguna lectura supera el umbral de temperatura
alertas_temp   = np.any(estacion[:, 0] > umbrales[0])
alertas_viento = np.any(estacion[:, 2] > umbrales[2])
print(f'¿Alerta temperatura (> {umbrales[0]})? {alertas_temp}')
print(f'¿Alerta viento (> {umbrales[2]})? {alertas_viento}')

# Cuántas horas en alerta por cada variable
for i, (var, umb) in enumerate(zip(variables, umbrales)):
    n = np.sum(estacion[:, i] > umb)
    print(f'{var}: {n} hora(s) superan umbral {umb}')
```

```python
# Paso 4: Normalización Z-score (estandarización)
media   = np.mean(estacion, axis=0)
desv    = np.std(estacion, axis=0)
normalizado = (estacion - media) / desv

print('=== Datos normalizados (Z-score) ===')
print(np.round(normalizado, 3))
print()
print('Media de normalizado (debe ser ~0):', np.round(np.mean(normalizado, axis=0), 10))
print('Std de normalizado (debe ser ~1): ', np.round(np.std(normalizado, axis=0), 10))
```

```python
# Paso 5: Visualización
horas = np.arange(8)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))
fig.suptitle('Análisis Meteorológico — 8 Horas de Medición', fontsize=13, fontweight='bold')

configs = [
    ('Temperatura (°C)',    '#2E75B6', umbrales[0]),
    ('Humedad (%)',         '#70AD47', umbrales[1]),
    ('Viento (km/h)',       '#ED7D31', umbrales[2]),
]

for i, (titulo, color, umbral) in enumerate(configs):
    axes[i].plot(horas, estacion[:, i], marker='o', color=color, linewidth=2)
    axes[i].axhline(umbral, color='red', linestyle='--', alpha=0.7, label=f'Umbral {umbral}')
    axes[i].set_title(titulo, fontweight='bold')
    axes[i].set_xlabel('Hora')
    axes[i].legend(fontsize=8)
    axes[i].grid(linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

## 📋 Resumen de operaciones y funciones

| Categoría | Operación / Función | Descripción |
|-----------|--------------------|--------------|
| **Indexación** | `arr[i]`, `arr[i,j]` | Elemento por posición |
| **Slicing** | `arr[a:b]`, `arr[:, col]` | Subconjunto de datos |
| **Condicional** | `arr[arr > x]` | Filtro booleano |
| **Combinado** | `arr[(cond1) & (cond2)]` | AND/OR/NOT lógico |
| **Copia** | `arr.copy()` | Independiente del original |
| **Aritmética** | `+`, `-`, `*`, `/`, `**`, `%` | Elemento a elemento |
| **Agregación** | `np.sum`, `np.mean`, `np.std` | Estadísticas globales o por eje |
| **Booleanas** | `np.any`, `np.all`, `np.sum` | Verificación de condiciones |
| **Matemáticas** | `np.sqrt`, `np.log`, `np.exp` | Transformaciones elementales |
| **Por eje** | `np.sum(m, axis=0/1)` | Por columnas (0) o filas (1) |

> 💡 **Regla de oro de las copias:** Siempre usa `.copy()` antes de transformar datos que no deben modificarse. Un alias silencioso puede propagar errores difíciles de detectar.

> 💡 **Broadcasting:** NumPy expande automáticamente dimensiones compatibles — pero siempre verifica `.shape` antes de combinar arreglos de distinto tamaño.
