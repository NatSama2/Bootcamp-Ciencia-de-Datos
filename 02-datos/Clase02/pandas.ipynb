```python
import pandas as pd 

data = {'Ciudad': ['Santiago', 'Valparaíso', 'Concepción'],
        'Producto': ['Libros', 'Ropa', 'Calzado'],
        'Costo': [4500, 6700, 8900],
        'Entregado': ['Si', 'No', 'Si']}

df = pd.DataFrame(data)
print(df)
```

```python
#Filtrar por multiples codiciones
print(df[(df['Costo'] > 5000) & (df['Entregado'] == 'Si')])
```

```python
data = {'ID Incidente': [2001, 2002, 2003, 2004, 2005],
        'Sucursal': ['Valparaíso', 'Santiago', 'La Serena', 'Temuco', 'Antofagasta'],
        'Prioridad': ['Alta', 'Media', 'Alta', 'Baja', 'Alta'],
        'Estado':['Pendiente', 'Resuelto', 'Pendiente', 'Resuelto', 'Pendiente']}
df = pd.DataFrame(data)

#Muestra el listado completo de incidentes.
print(df)
```

```python
# Selecciona todas las sucursales afectadas por incidentes de prioridad "Alta".
filtro_prioridad_alta = df[df['Prioridad'] == 'Alta']
print(filtro_prioridad_alta)



```

```python
# Filtra los registros en los que: La prioridad sea "Alta" y El estado sea "Pendiente"
filtro_prioridad_alta_pendiente = df[(df['Prioridad'] == 'Alta') & (df['Estado'] == 'Pendiente')]
print(filtro_prioridad_alta_pendiente)
```

```python
# ¿Cuántos incidentes críticos sin resolver hay actualmente?
filtro_criticos_sin_resolver = df[(df['Prioridad'] == 'Alta') & (df['Estado'] == 'Pendiente')]
cantidad_criticos_sin_resolver = filtro_criticos_sin_resolver.shape[0]
print(f"Cantidad de incidentes críticos sin resolver: {cantidad_criticos_sin_resolver}")
```

```python
# Extrae los valores de Sucursal correspondientes solo a los casos filtrados anteriormente.
sucursales_criticos_sin_resolver = filtro_criticos_sin_resolver['Sucursal']
print(filtro_criticos_sin_resolver['Sucursal'])
```

```python
import pandas as pd

# 1. Creación del DataFrame
datos = {
    'Producto': ['Paracetamol', 'Ibuprofeno', 'Amoxicilina', 'Aspirina', 'Loratadina'],
    'Precio': [1200, 1500, 2500, 900, 1800],
    'Stock': [35, 20, 0, 50, 40],
    'Farmacia': ['Central', 'Norte', 'Sur', 'Central', 'Oriente']
}
df = pd.DataFrame(datos)
print(df)
```

```python
# 2. Exploración Rápida
print("--- HEAD ---")
print(df.head())

print("\n--- TAIL ---")
print(df.tail())

print("\n--- INFO ---")
print(df.info())

print("\n--- DESCRIBE ---")
print(df.describe())
```

```python
import pandas as pd
data = {'Nombre': ['Valentina', 'Pedro', 'Laura', 'Javiera', 'Carlos', 'Martín'],
        'Edad': [29, 35, 31, 28, 42, 38],
        'Experiencia (años)': [3, 8, 5, 2, 10, 6],
        'Departamento': ['TI', 'TI', 'RRHH', 'TI', 'RRHH', 'Ventas']}
df = pd.DataFrame(data)

print("Edad mínima:", df['Edad'].min())
print("Edad máxima:", df['Edad'].max())
print("Edad promedio:", df['Edad'].mean())
print("Edad mediana:", df['Edad'].median())
print("Total de años de experiencia:", df['Experiencia (años)'].sum())
print("Cantidad de datos válidos:", df['Edad'].count())
print("Departamentos únicos:", df['Departamento'].unique())
print("Frecuencia de departamentos:")
print(df['Departamento'].value_counts())
```
