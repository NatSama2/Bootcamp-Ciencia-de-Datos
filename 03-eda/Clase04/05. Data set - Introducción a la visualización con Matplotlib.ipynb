## PARTE 2 — Interfaz procedural vs Orientada a Objetos

### 2.1 Interfaz procedural (rápida, para tareas simples)

```python
# Código exacto de la presentación — slide 10 (gráfico de barras)
import matplotlib.pyplot as plt

meses   = ['Ene', 'Feb', 'Mar', 'Abr']
ventas  = [100, 120, 90, 150]

plt.bar(meses, ventas)
plt.title('Ventas por mes')
plt.xlabel('Mes')
plt.ylabel('Monto en miles')
plt.grid(True)
plt.show()
```

### 2.2 Interfaz orientada a objetos (recomendada — control total)

```python
# Código exacto de la presentación — slide 18
import matplotlib.pyplot as plt

# Crear figura y eje
fig, ax = plt.subplots()

# Trazar línea
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], label='Ventas')

# Personalización del gráfico
ax.set_title('Evolución de ventas')
ax.set_xlabel('Trimestre')
ax.set_ylabel('Miles de unidades')
ax.legend()
ax.grid(True)

# Mostrar figura
plt.show()
```

```python
# Comparación directa: procedural vs OOP
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
fig.suptitle('Interfaz Procedural vs Orientada a Objetos', fontweight='bold')

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# ── Interfaz procedural (izquierda) ──
plt.sca(axes[0])       # seleccionar el eje
plt.plot(x, y, 'b-o')
plt.title('Interfaz procedural\nplt.plot(), plt.title()...')
plt.xlabel('Trimestre')
plt.ylabel('Miles')
plt.grid(True)

# ── Interfaz OOP (derecha) ──
axes[1].plot(x, y, 'r-s', label='Ventas')
axes[1].set_title('Interfaz OOP\nfig, ax = plt.subplots()')
axes[1].set_xlabel('Trimestre')
axes[1].set_ylabel('Miles')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

print('Recomendación: Usa fig, ax = plt.subplots() para proyectos con múltiples gráficos.')
```

## PARTE 3 — Anatomía de un gráfico en Matplotlib

### 3.1 Los componentes clave

```python
# Tabla anatómica de la presentación — slide 17
tabla_anat = pd.DataFrame({
    'Elemento': ['Figure','Axes','Axis','Tick','Label','Title','Legend','Grid'],
    'Descripción': [
        'Objeto contenedor global. Puede incluir varios ejes (Axes).',
        'Área de trazado donde se colocan datos, etiquetas, título, etc.',
        'Los ejes X o Y en sí mismos, contienen ticks y valores numéricos.',
        'Marcadores en los ejes, indican valores individuales.',
        'Etiquetas asociadas a cada eje (xlabel, ylabel).',
        'Título general del gráfico o por cada Axes.',
        'Explicación de lo que representa cada serie de datos.',
        'Líneas auxiliares que mejoran la legibilidad numérica.'
    ]
})
print(tabla_anat.to_string(index=False))
```

```python
# Gráfico anotado que muestra cada componente visualmente
fig, ax = plt.subplots(figsize=(10, 6))

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [8, 15, 22, 28, 35]

ax.plot(x, y1, 'b-o', linewidth=2, label='Serie A')
ax.plot(x, y2, 'r--s', linewidth=2, label='Serie B')

# ── COMPONENTES ANOTADOS ──
ax.set_title('TITLE: Título del gráfico', fontsize=13, fontweight='bold', color='darkblue')
ax.set_xlabel('LABEL (xlabel): Etiqueta eje X', fontsize=10, color='purple')
ax.set_ylabel('LABEL (ylabel): Etiqueta eje Y', fontsize=10, color='purple')
ax.legend(title='LEGEND', fontsize=9, title_fontsize=9)
ax.grid(True, linestyle='--', alpha=0.5)

# Anotaciones
ax.annotate('← AXES (área de dibujo)', xy=(3, 22), xytext=(3.5, 15),
            arrowprops=dict(arrowstyle='->', color='green'),
            color='green', fontsize=9)
ax.annotate('TICK →', xy=(2, 0), xytext=(2.2, -3),
            fontsize=8, color='brown')

ax.text(0.02, 0.98, 'FIGURE (contenedor global)',
        transform=fig.transFigure, fontsize=8, color='gray',
        verticalalignment='top')

# Borde de la figura
for spine in ['top','right','bottom','left']:
    ax.spines[spine].set_linewidth(2)
    ax.spines[spine].set_edgecolor('#CCCCCC')

plt.tight_layout()
plt.show()
```

## PARTE 4 — Galería de tipos de gráficos con pyplot

### 4.1 Los 5 tipos básicos

```python
np.random.seed(42)

fig, axes = plt.subplots(2, 3, figsize=(16, 9))
fig.suptitle('Galería de gráficos básicos en Matplotlib', fontweight='bold', fontsize=13)

# 1. plot() — línea
x = [1, 2, 3, 4, 5]
axes[0,0].plot(x, [10,20,25,30,40], 'b-o', linewidth=2, label='Serie A')
axes[0,0].plot(x, [15,18,22,26,35], 'r--s', linewidth=2, label='Serie B')
axes[0,0].set_title('plot() — Línea')
axes[0,0].set_xlabel('Semana')
axes[0,0].set_ylabel('Unidades')
axes[0,0].legend()
axes[0,0].grid(True)

# 2. bar() — barras verticales
meses  = ['Ene','Feb','Mar','Abr']
ventas = [100, 120, 90, 150]
axes[0,1].bar(meses, ventas, color='#2E75B6', edgecolor='white')
axes[0,1].set_title('bar() — Barras verticales')
axes[0,1].set_xlabel('Mes')
axes[0,1].set_ylabel('Monto en miles')
axes[0,1].grid(True, axis='y')

# 3. barh() — barras horizontales
regiones = ['Norte','Sur','Centro','Este','Oeste']
valores  = [45, 38, 62, 55, 41]
axes[0,2].barh(regiones, valores, color='#70AD47', edgecolor='white')
axes[0,2].set_title('barh() — Barras horizontales')
axes[0,2].set_xlabel('Entregas')
axes[0,2].grid(True, axis='x')

# 4. scatter() — dispersión
x_s = np.random.randint(10, 100, 50)
y_s = x_s * 0.8 + np.random.normal(0, 10, 50)
axes[1,0].scatter(x_s, y_s, color='#ED7D31', alpha=0.6, s=40)
axes[1,0].set_title('scatter() — Dispersión')
axes[1,0].set_xlabel('Variable X')
axes[1,0].set_ylabel('Variable Y')
axes[1,0].grid(True)

# 5. hist() — histograma
datos = np.random.normal(50, 12, 200)
axes[1,1].hist(datos, bins=15, color='#7030A0', edgecolor='white', alpha=0.8)
axes[1,1].set_title('hist() — Histograma')
axes[1,1].set_xlabel('Valor')
axes[1,1].set_ylabel('Frecuencia')
axes[1,1].grid(True, axis='y')

# 6. pie() — torta
labels_p = ['Zona A','Zona B','Zona C','Zona D']
sizes_p  = [35, 25, 20, 20]
axes[1,2].pie(sizes_p, labels=labels_p, autopct='%1.1f%%',
              startangle=90, colors=['#2E75B6','#ED7D31','#70AD47','#FFC000'])
axes[1,2].set_title('pie() — Gráfico de torta')

plt.tight_layout()
plt.show()
```

## 📋 Resumen de componentes y buenas prácticas

### Anatomía del gráfico

| Componente | Descripción | Método OOP |
|-----------|-------------|------------|
| `Figure` | Contenedor global de la visualización | `fig, ax = plt.subplots()` |
| `Axes` | Área de trazado (¡no confundir con Axis!) | `ax = ...` |
| `Axis` | Los ejes X e Y con ticks y valores | `ax.xaxis`, `ax.yaxis` |
| `Title` | Título del gráfico | `ax.set_title('...')` |
| `Label` | Etiquetas de ejes | `ax.set_xlabel('...')` |
| `Legend` | Leyenda para identificar series | `ax.legend()` |
| `Grid` | Cuadrícula de referencia | `ax.grid(True)` |

### Procedural vs OOP

| | Procedural (`plt.*`) | OOP (`fig, ax`) |
|-|---------------------|-------------------|
| Sintaxis | `plt.plot()`, `plt.title()` | `ax.plot()`, `ax.set_title()` |
| Cuándo usar | Scripts rápidos, exploratorios | Proyectos con múltiples subgráficos |
| Control | Limitado | Total |
| Recomendado para | Aprendizaje inicial | Producción |

> 💡 **`fig, ax = plt.subplots()`** es la forma recomendada para tener control total. Úsala siempre en proyectos que van más allá de un gráfico rápido.

> 💡 **`axes` ≠ `axis`:** `axes` es el área de dibujo (el subplot completo). `axis` son las líneas X o Y con sus ticks. Confundirlos es uno de los errores más comunes.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Paso 2: Datos
semanas = ["Semana 1", "Semana 2", "Semana 3", "Semana 4"]
entregas_zona_norte = [120, 135, 150, 160]
entregas_zona_sur = [100, 110, 95, 130]

# Paso 3: Generar la visualización
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(semanas, entregas_zona_norte, label="Zona Norte", marker='o')
ax.plot(semanas, entregas_zona_sur, label="Zona Sur", marker='s')

ax.set_title("Entregas semanales por zona")
ax.set_xlabel("Semana")
ax.set_ylabel("Cantidad de entregas")
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()

# Paso 4: Guardar el gráfico
fig.savefig("entregas_semanales.png")
```
