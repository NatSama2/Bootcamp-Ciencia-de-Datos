```python
# Instalación (si es necesario):
# En terminal:     pip install seaborn
# En Google Colab: !pip install seaborn

# Importación correcta — código exacto de la presentación
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Configurar estilo visual — buena práctica desde el inicio
sns.set_theme(style='whitegrid')
plt.rcParams['figure.dpi'] = 110

print('✅ Seaborn importado correctamente')
print(f'Versión de Seaborn: {sns.__version__}')
```

### 1.2 Seaborn vs Matplotlib — ejemplo comparativo

```python
# Código exacto de la presentación — comparación
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle('Seaborn vs Matplotlib — mismo gráfico, diferente código', fontweight='bold')

# Con Matplotlib:
axes[0].plot([1, 2, 3], [4, 5, 6])
axes[0].set_title('Con Matplotlib')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

# Con Seaborn:
sns.lineplot(x=[1, 2, 3], y=[4, 5, 6], ax=axes[1])
axes[1].set_title('Con Seaborn')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

plt.tight_layout()
plt.show()

print('Seaborn: menos código, estilo integrado, mejor presentación predeterminada.')
```

### 1.3 Estilos disponibles en Seaborn

```python
estilos = ['white', 'whitegrid', 'dark', 'darkgrid', 'ticks']
np.random.seed(42)
datos = np.random.normal(50, 15, 100)

fig, axes = plt.subplots(1, 5, figsize=(18, 3))
fig.suptitle('Estilos disponibles en sns.set_theme(style=...)', fontweight='bold')

for ax, estilo in zip(axes, estilos):
    with sns.axes_style(estilo):
        ax.hist(datos, bins=15, color='#2E75B6', edgecolor='white')
        ax.set_title(f'style="{estilo}"', fontsize=9)
        ax.set_xlabel('')

plt.tight_layout()
plt.show()
print('Recomendado para análisis profesional: "whitegrid"')
```

## PARTE 2 — Clasificación de gráficos

### 2.1 Tabla de gráficos por propósito (de la presentación)

```python
tabla_graficos = pd.DataFrame({
    'Propósito':      ['Distribución','Distribución','Comparación','Comparación',
                       'Relación','Relación','Composición','Composición',
                       'Multivariado','Multivariado'],
    'Tipo de gráfico':['Histograma','Gráfico de densidad','Boxplot','Violin plot',
                       'Dispersión (scatter)','Regresión lineal','Barras','Conteo',
                       'Gráfico de pares','Mapa de calor'],
    'Función Seaborn': ['sns.histplot()','sns.kdeplot()','sns.boxplot()',
                        'sns.violinplot()','sns.scatterplot()','sns.regplot()',
                        'sns.barplot()','sns.countplot()',
                        'sns.pairplot()','sns.heatmap()'],
    'Variables':       ['Numérica','Numérica','Num + Cat','Num + Cat',
                        'Num + Num','Num + Num','Cat + Num','Categórica',
                        'Múltiples','Correlación']
})
print(tabla_graficos.to_string(index=False))
```

### 2.2 Galería visual de todos los tipos

```python
# Datos de demostración
np.random.seed(42)
df_demo = pd.DataFrame({
    'valor':     np.random.normal(50, 12, 120),
    'grupo':     np.random.choice(['A','B','C'], 120),
    'variable2': np.random.normal(30, 8, 120),
    'categoria': np.random.choice(['X','Y'], 120)
})

fig, axes = plt.subplots(2, 4, figsize=(18, 8))
fig.suptitle('Galería de tipos de gráficos en Seaborn', fontweight='bold', fontsize=13)

# 1. histplot
sns.histplot(df_demo['valor'], bins=15, kde=True, ax=axes[0,0], color='#2E75B6')
axes[0,0].set_title('histplot\n(Distribución numérica)', fontsize=9)

# 2. kdeplot
for g, col in zip(['A','B','C'], ['#2E75B6','#ED7D31','#70AD47']):
    sns.kdeplot(df_demo[df_demo['grupo']==g]['valor'], ax=axes[0,1],
                color=col, label=g, linewidth=2)
axes[0,1].set_title('kdeplot\n(Densidad por grupo)', fontsize=9)
axes[0,1].legend(fontsize=7)

# 3. boxplot
sns.boxplot(data=df_demo, x='grupo', y='valor', ax=axes[0,2], palette='Set2')
axes[0,2].set_title('boxplot\n(Comparación por categoría)', fontsize=9)

# 4. violinplot
sns.violinplot(data=df_demo, x='grupo', y='valor', ax=axes[0,3], palette='Set2')
axes[0,3].set_title('violinplot\n(Distribución + comparación)', fontsize=9)

# 5. scatterplot
sns.scatterplot(data=df_demo, x='variable2', y='valor', hue='grupo',
                ax=axes[1,0], palette='Set1', s=40, alpha=0.7)
axes[1,0].set_title('scatterplot\n(Relación num-num)', fontsize=9)

# 6. barplot
sns.barplot(data=df_demo, x='grupo', y='valor', ax=axes[1,1],
            palette='Blues_r', errorbar='sd')
axes[1,1].set_title('barplot\n(Promedio por categoría)', fontsize=9)

# 7. countplot
sns.countplot(data=df_demo, x='grupo', hue='categoria', ax=axes[1,2], palette='Set2')
axes[1,2].set_title('countplot\n(Conteo categórico)', fontsize=9)

# 8. heatmap
corr = df_demo[['valor','variable2']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            ax=axes[1,3], vmin=-1, vmax=1)
axes[1,3].set_title('heatmap\n(Correlación)', fontsize=9)

plt.tight_layout()
plt.show()
```

### 3.3 Gráfico b — Relación entre edad y horas de visualización (`scatterplot`)

```python
plt.figure(figsize=(9, 4))
sns.scatterplot(data=df, x='edad', y='horas_visualizacion',
                hue='suscripcion_activa',
                palette={'Sí': '#2E75B6', 'No': '#ED7D31'},
                s=60, alpha=0.7)
plt.title('Relación entre edad y horas de visualización', fontweight='bold')
plt.xlabel('Edad')
plt.ylabel('Horas por semana')
plt.legend(title='Suscripción activa')
plt.tight_layout()
plt.show()
```

```python
print('=== Promedio y std por dispositivo ===')
print(df.groupby('dispositivo')['horas_visualizacion'].agg(['mean','std','count']).round(2))

# ✏️ ¿Qué dispositivo muestra mayor uso? ¿Qué estrategia recomendarías?
obs_c = ""
print(f'Observación: {obs_c}')
```

## PARTE 4 — Actividad autónoma: Patrones de compra en clientes

### 4.1 Cargar el dataset

```python
# Código exacto de la presentación
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df2 = pd.read_csv('ventas_clientes_autonomo.csv')
print(df2.head(8))
print()
print(df2.dtypes)
```

## 📋 Resumen de tipos de gráficos y cuándo usarlos

| Función | Propósito | Variables | ¿Cuándo usarla? |
|---------|-----------|-----------|----------------|
| `sns.histplot()` | Distribución | 1 numérica | Ver forma, sesgo, outliers |
| `sns.kdeplot()` | Distribución | 1 numérica | Comparar distribuciones entre grupos |
| `sns.boxplot()` | Comparación | Num + Cat | Detectar outliers y comparar medianas |
| `sns.violinplot()` | Comparación | Num + Cat | Ver distribución completa por grupo |
| `sns.scatterplot()` | Relación | Num + Num | Ver asociación entre dos variables |
| `sns.regplot()` | Relación | Num + Num | Visualizar tendencia lineal |
| `sns.barplot()` | Composición | Cat + Num | Comparar promedios por grupo |
| `sns.countplot()` | Composición | Categórica | Frecuencia de categorías |
| `sns.pairplot()` | Multivariado | Múltiples | Explorar relaciones entre varias vars |
| `sns.heatmap()` | Multivariado | Correlación | Visualizar matriz de correlación |

> 💡 **Regla práctica:** Antes de elegir el gráfico, responde: ¿Cuántas variables? ¿De qué tipo? ¿Quiero ver distribución, comparación, relación o composición?

> 💡 **Siempre incluir:** título descriptivo, etiquetas de ejes X e Y, leyenda si hay `hue`, y verificar que los datos estén limpios antes de graficar.
