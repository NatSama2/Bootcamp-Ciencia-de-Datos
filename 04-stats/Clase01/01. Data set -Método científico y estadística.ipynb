### 3.1 Cargar el dataset — código exacto del slide 33

```python
# Código exacto de la presentación — slide 33
import pandas as pd

df = pd.read_csv('abandono_escolar.csv')  # Contiene: comuna, nivel socioeconómico, asistencia, conectividad, etc.
print(df.head())
print(df.info())
```

```python
print('=== Estadísticas descriptivas ===')
print(df.describe().round(2))
print()
print('=== Distribución de abandono ===')
print(df['abandono'].value_counts())
print(f'Tasa de abandono: {df["abandono"].mean()*100:.1f}%')
```

```python
# Exploración inicial
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
fig.suptitle('Exploración inicial — Abandono escolar', fontweight='bold', fontsize=13)

# Tasa de abandono por NSE
tasa_nse = df.groupby('nivel_socioeconomico')['abandono'].mean() * 100
orden_nse = ['Bajo','Medio-bajo','Medio','Medio-alto']
tasa_nse = tasa_nse.reindex(orden_nse)
axes[0,0].bar(tasa_nse.index, tasa_nse.values, color='#ED7D31', edgecolor='white')
axes[0,0].set_title('Tasa de abandono por NSE')
axes[0,0].set_ylabel('% abandono')
axes[0,0].tick_params(axis='x', rotation=15)
axes[0,0].grid(True, axis='y', alpha=0.4)

# Distribución asistencia por abandono
for abandono, color, label in [(0,'#2E75B6','No abandono'), (1,'#FC4E4E','Abandono')]:
    sns.kdeplot(df[df['abandono']==abandono]['pct_asistencia'],
                ax=axes[0,1], fill=True, alpha=0.4, color=color, label=label)
axes[0,1].set_title('Distribución de asistencia por abandono')
axes[0,1].set_xlabel('% asistencia')
axes[0,1].legend()

# Scatter: conectividad vs asistencia
axes[1,0].scatter(df[df['abandono']==0]['pct_conectividad'],
                   df[df['abandono']==0]['pct_asistencia'],
                   color='#2E75B6', alpha=0.4, s=25, label='No abandono')
axes[1,0].scatter(df[df['abandono']==1]['pct_conectividad'],
                   df[df['abandono']==1]['pct_asistencia'],
                   color='#FC4E4E', alpha=0.5, s=25, label='Abandono')
axes[1,0].set_title('Conectividad vs Asistencia')
axes[1,0].set_xlabel('% conectividad')
axes[1,0].set_ylabel('% asistencia')
axes[1,0].legend(fontsize=8)

# Promedio notas por abandono
sns.boxplot(data=df, x='abandono', y='promedio_notas',
            palette={'0':'#2E75B6', '1':'#FC4E4E'}, ax=axes[1,1])
axes[1,1].set_xticklabels(['No abandono','Abandono'])
axes[1,1].set_title('Distribución de notas por abandono')
axes[1,1].set_ylabel('Promedio de notas')
axes[1,1].axhline(4.0, color='red', linestyle='--', linewidth=1.5, label='Mín. aprobación')
axes[1,1].legend(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Correlaciones con la variable abandono
print('=== Correlación con variable abandono ===')
vars_num = ['pct_asistencia','pct_conectividad','promedio_notas']
for v in vars_num:
    r = df[v].corr(df['abandono'])
    print(f'  {v:<25}: r = {r:.3f}')

print()
print('Nota: correlaciones negativas sugieren que a mayor asistencia/conectividad/notas,\n',
      '      menor probabilidad de abandono. Estas serán las bases de nuestra hipótesis.')
```

## 📋 Resumen de conceptos clave

### Fundamentos filosóficos

| Concepto | Autor/Corriente | Relevancia en datos |
|---------|----------------|--------------------|
| Epistemología | Filosofía general | Justificar el conocimiento generado |
| Positivismo lógico | Escuela de Viena (Carnap) | Todo enunciado debe ser verificable empíricamente |
| Falsabilidad | Karl Popper | Si no puede refutarse, no es científico |

### Las 6 características del método científico

| Característica | Pregunta clave | En Python/datos |
|--------------|---------------|----------------|
| Facticidad | ¿Está basado en datos reales? | Usar fuentes verificables |
| Objetividad | ¿Puede otro analista replicarlo? | Código reproducible |
| Racionalidad | ¿La inferencia es lógicamente válida? | No asumir causalidad por correlación |
| Sistematicidad | ¿Se siguió un orden metodológico? | Pipeline documentado |
| Reproducibilidad | ¿Otros obtienen el mismo resultado? | `np.random.seed()`, notebooks compartidos |
| Refutabilidad | ¿La hipótesis puede probarse falsa? | Métricas explícitas en conjunto de prueba |

### Buenas prácticas (slide 20)

| Práctica | Justificación |
|---------|---------------|
| Formular hipótesis antes de explorar datos | Evita conclusiones basadas solo en correlaciones aparentes |
| Controlar variables externas | Mejora la validez interna y reduce el sesgo |
| Documentar todo el proceso | Aumenta reproducibilidad y permite auditoría |
| Validar con datos nuevos | Evita sobreajuste |
| Compartir código y metodología | Fomenta transparencia |

> 💡 **Falsabilidad práctica:** antes de entregar un modelo, hazte la pregunta de Popper: *¿Qué resultado me haría concluir que este modelo no funciona?* Si no hay respuesta posible, el modelo no está siendo evaluado científicamente.
