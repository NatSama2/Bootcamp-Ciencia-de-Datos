# **1. ¿Qué es son las sentencias iterativas y por qué se necesitan? 🎯**

Las sentencias iterativas son bloques fundamentales en programación que nos permiten ejecutar código repetidamente. Imagina que son como un ciclo que se repite hasta que se cumple una condición específica.

Hay dos tipos principales:

1. while: Repite mientras una condición sea verdadera.
2. for: Repite un número específico de veces o sobre una colección de elementos.

Pensemos en un cocinero preparando órdenes en un restaurante:

* Escenario 1 (while): "Mientras haya clientes en el restaurante, sigue cocinando"
* Escenario 2 (for): "Para cada orden en la lista, prepara un plato"

En Programación 💻

```python
# Ejemplo 1: while (mientras haya clientes)
clientes = 5
while clientes > 0:
    print(f"Atendiendo al cliente {clientes}")
    clientes -= 1

# Ejemplo 2: for (para cada orden)
ordenes = ["🍔", "🍕", "🌮", "🍜"]
for orden in ordenes:
    print(f"Preparando: {orden}")
```

# **2. La Sentencia while ⭐**

while es una estructura que repite un bloque de código mientras una condición sea verdadera. Es como un guardia que verifica una condición antes de permitir que el código se ejecute nuevamente.

**Características clave:**

* La condición se verifica ANTES de cada iteración.
* El código se ejecutará 0 o más veces.
* Necesitas asegurarte de que la condición eventualmente sea falsa.

Imagina una máquina expendedora de bebidas:

1. Verifica si quedan refrescos (condición).
2. Si hay refrescos, entrega uno (acción).
3. Actualiza el inventario (actualización).
4. Vuelve al paso 1.

💻 Ejemplo Práctico Detallado:

```python
# Simulador de máquina expendedora
refrescos = 5
dinero_recolectado = 0

print("🥤 Máquina Expendedora")
print("=" * 30)

while refrescos > 0:
    print(f"\nRefrescos disponibles: {'🥤' * refrescos}")
    print(f"Dinero recolectado: ${dinero_recolectado}")

    # Simular venta
    venta = input("¿Desea comprar un refresco? (s/n): ")
    if venta.lower() == 's':
        refrescos -= 1
        dinero_recolectado += 2
        print("¡Gracias por su compra! 🎉")
    else:
        break

print("\n=== Estado Final ===")
print(f"Refrescos restantes: {refrescos}")
print(f"Dinero total: ${dinero_recolectado}")
```

# **🎮 Ejercicio Guiado: Juego de Piedra, Papel o Tijeras**

Objetivo: Crear un juego interactivo donde el usuario juega contra la computadora hasta que decida detenerse.

Instrucciones:

1. Primero, importamos las bibliotecas necesarias:

```python
import random
```

2. Definimos las opciones del juego:

```python
opciones = ["piedra", "papel", "tijeras"]
puntaje_jugador = 0
puntaje_computadora = 0
```

3. Creamos la lógica del juego:

```python
def determinar_ganador(jugador, computadora):
    if jugador == computadora:
        return "Empate"
    elif ((jugador == "piedra" and computadora == "tijeras") or
          (jugador == "papel" and computadora == "piedra") or
          (jugador == "tijeras" and computadora == "papel")):
        return "Jugador"
    else:
        return "Computadora"
```

4. Implementamos el bucle principal:

```python
jugando = True

while jugando:
    # Mostrar marcador
    print("\n=== Marcador ===")
    print(f"Jugador: {puntaje_jugador} 👤")
    print(f"Computadora: {puntaje_computadora} 💻")

    # Obtener jugadas
    jugada_usuario = input("\n¿Piedra, papel o tijeras? ").lower()
    jugada_computadora = random.choice(opciones)

    print(f"\nComputadora eligió: {jugada_computadora}")

    # Determinar ganador
    resultado = determinar_ganador(jugada_usuario, jugada_computadora)

    if resultado == "Jugador":
        print("¡Ganaste! 🎉")
        puntaje_jugador += 1
    elif resultado == "Computadora":
        print("¡Perdiste! 😢")
        puntaje_computadora += 1
    else:
        print("¡Empate! 🤝")

    # Preguntar si quiere seguir jugando
    seguir = input("\n¿Quieres seguir jugando? (s/n): ")
    if seguir.lower() != 's':
        jugando = False

print("\n=== Resultado Final ===")
print(f"Jugador: {puntaje_jugador} 👤")
print(f"Computadora: {puntaje_computadora} 💻")
```

🎯 ¡Manos a la obra! - Simulador de Banco

Objetivo: Crear un programa que simule un cajero automático con las siguientes características:

* Mantener un saldo,
* Permitir depósitos y retiros,
* Mostrar el saldo actual,
* Continuar hasta que el usuario decida salir.

**Instrucciones:**

1. Crear variables para:
  * Saldo inicial.
  * Estado de la operación.
2. Crear un menú con opciones:
  * Ver saldo
  * Depositar
  * Retirar
  * Salir
3. Implementar la lógica para cada operación:
  * Validar montos.
  * Actualizar saldo.
  * Mostrar mensajes apropiados.

**Tu código aquí:**

```python
# Inicialización de variables
saldo = 1000
operando = True

def ver_saldo(saldo):
    print(f"\n Tu saldo actual es: ${saldo:.2f}")

def depositar(saldo):
    monto = float(input("¿Cuánto deseas depositar? $"))
    if monto <= 0:
        print(" El monto debe ser mayor a 0.")
        return saldo
    saldo += monto
    print(f"✅ Depositaste ${monto:.2f}. Nuevo saldo: ${saldo:.2f}")
    return saldo

def retirar(saldo):
    monto = float(input("¿Cuánto deseas retirar? $"))
    if monto <= 0:
        print(" El monto debe ser mayor a 0.")
        return saldo
    elif monto > saldo:
        print(f" Fondos insuficientes. Tu saldo es ${saldo:.2f}.")
        return saldo
    saldo -= monto
    print(f" Retiraste ${monto:.2f}. Nuevo saldo: ${saldo:.2f}")
    return saldo

while operando:
    # Mostrar menú
    print("\n=== Cajero Automático  ===")
    print("1. Ver saldo")
    print("2. Depositar")
    print("3. Retirar")
    print("4. Salir")

    opcion = input("\nElige una opción (1-4): ")

    if opcion == "1":
        ver_saldo(saldo)
    elif opcion == "2":
        saldo = depositar(saldo)
    elif opcion == "3":
        saldo = retirar(saldo)
    elif opcion == "4":
        operando = False
    else:
        print(" Opción no válida. Intenta de nuevo.")

print("\n=== Sesión Finalizada ===")
print(f"Saldo final: ${saldo:.2f} ")
print("¡Gracias por usar el cajero! ")
```

# **3. La Sentencia for: El Iterador por Excelencia 🔄**

La sentencia for es una estructura que nos permite iterar sobre una secuencia de elementos de manera controlada. A diferencia de while, que continúa hasta que una condición sea falsa, for repetirá el código un número específico de veces o para cada elemento en una colección.

**Características clave:**

* Ideal para colecciones conocidas.
* Mantiene un control automático del índice.
* Garantiza que visitaremos cada elemento una vez.
* No requiere una condición de salida manual.

Imagina un cartero repartiendo correo:
* Tiene una bolsa con cartas (la colección).
* Para cada carta en la bolsa:
  * Lee la dirección.
  * Entrega la carta.
  * Pasa a la siguiente.
  * Se detiene cuando no quedan más cartas.

💻 Observa el siguiente ejemplo:

```python
# Simulador de entrega de correo
cartas = ["📫 Factura", "💌 Carta personal", "📦 Paquete", "🏷️ Postal"]

print("🚚 Simulador de Entrega de Correo")
print("=" * 35)

for indice, carta in enumerate(cartas, 1):
    print(f"\nEntrega #{indice}")
    print(f"Entregando: {carta}")
    input("Presiona Enter para continuar...")
    print(f"✅ {carta} entregado correctamente")

print("\n📮 ¡Todas las entregas completadas!")
```

# **🎮 Ejercicio Guiado: Sistema de Calificaciones**

Objetivo: Crear un sistema que procese las calificaciones de un grupo de estudiantes y genere estadísticas.

Instrucciones:

1. Primero, configuramos los datos:

```python
estudiantes = {
    "Ana": [90, 85, 88],
    "Juan": [75, 82, 89],
    "María": [95, 91, 93],
    "Carlos": [70, 75, 77]
}
```

2. Creamos una función para calcular el promedio:

```python
def calcular_promedio(notas):
    return sum(notas) / len(notas)
```

3. Procesamos cada estudiante:

```python
print("📊 REPORTE DE CALIFICACIONES")
print("=" * 30)

for estudiante, notas in estudiantes.items():
    # Calculamos estadísticas
    promedio = calcular_promedio(notas)
    nota_maxima = max(notas)
    nota_minima = min(notas)

    # Mostramos resultados
    print(f"\n👤 Estudiante: {estudiante}")
    print(f"📝 Notas: {notas}")
    print(f"📈 Promedio: {promedio:.2f}")
    print(f"⭐ Nota más alta: {nota_maxima}")
    print(f"📉 Nota más baja: {nota_minima}")

    # Determinamos estado
    if promedio >= 90:
        print("🏆 ¡Excelente desempeño!")
    elif promedio >= 80:
        print("✨ ¡Buen trabajo!")
    else:
        print("💪 ¡Sigue esforzándote!")
```

# **🎯 ¡Manos a la obra! - Sistema de Inventario de Videojuegos**

Objetivo: Crear un sistema que gestione un inventario de videojuegos, mostrando información detallada y estadísticas.

Instrucciones:

1. Crea un diccionario de videojuegos con la siguiente información para cada juego:
  * Título
  * Género
  * Precio
  * Stock
  * Valoración (1-5)
2. Implementa funciones para:
  * Mostrar todos los juegos.
  * Filtrar por género.
  * Encontrar juegos en oferta.
  * Calcular valor total del inventario.
3. Utiliza for para:
  * Recorrer el inventario.
  * Mostrar información formateada.
  * Calcular estadísticas.

**Tu código aquí:**

```python
# 🎮 Sistema de gestión de inventario de videojuegos

# 1. Diccionario de videojuegos
inventario_juegos = {
    "The Legend of Adventure": {
        "género": "RPG",
        "precio": 59.99,
        "stock": 10,
        "valoración": 4.8
    },
    "Speed Racers X": {
        "género": "Carreras",
        "precio": 39.99,
        "stock": 5,
        "valoración": 4.2
    },
    "Galaxy Warriors": {
        "género": "Acción",
        "precio": 49.99,
        "stock": 0,
        "valoración": 4.6
    },
    "Mystery Manor": {
        "género": "Aventura",
        "precio": 24.99,
        "stock": 15,
        "valoración": 3.9
    },
    "Dragon Quest Legends": {
        "género": "RPG",
        "precio": 19.99,
        "stock": 8,
        "valoración": 4.5
    },
    "Football Pro 2025": {
        "género": "Deportes",
        "precio": 29.99,
        "stock": 20,
        "valoración": 3.5
    }
}


# 2. Funciones del sistema

def mostrar_inventario():
    """Muestra todos los juegos con su información formateada."""
    print("INVENTARIO DE VIDEOJUEGOS")
    print("=" * 35)

    for titulo, datos in inventario_juegos.items():
        # Indicamos disponibilidad según el stock
        disponibilidad = "Disponible" if datos["stock"] > 0 else "Agotado"

        print(f"\n{titulo}")
        print(f"Género: {datos['género']}")
        print(f"Precio: ${datos['precio']:.2f}")
        print(f"Stock: {datos['stock']} ({disponibilidad})")
        print(f"Valoración: {datos['valoración']}/5")


def filtrar_por_genero(genero):
    """Muestra solo los juegos que pertenecen al género indicado."""
    print(f"\nJUEGOS DEL GÉNERO: {genero.upper()}")
    print("=" * 35)

    encontrados = 0
    for titulo, datos in inventario_juegos.items():
        if datos["género"].lower() == genero.lower():
            print(f"{titulo} — ${datos['precio']:.2f}")
            encontrados += 1

    if encontrados == 0:
        print("No se encontraron juegos de ese género.")


def juegos_en_oferta(precio_maximo=30.00):
    """Muestra los juegos cuyo precio sea menor o igual al precio máximo."""
    print(f"\nJUEGOS EN OFERTA (≤ ${precio_maximo:.2f})")
    print("=" * 35)

    hay_ofertas = False
    for titulo, datos in inventario_juegos.items():
        if datos["precio"] <= precio_maximo:
            print(f"{titulo} — ${datos['precio']:.2f}")
            hay_ofertas = True

    if not hay_ofertas:
        print("No hay juegos en oferta por ahora.")


def calcular_valor_total():
    """Calcula el valor total del inventario (precio × stock de cada juego)."""
    valor_total = 0
    for datos in inventario_juegos.values():
        valor_total += datos["precio"] * datos["stock"]
    return valor_total


def mostrar_estadisticas():
    """Calcula y muestra estadísticas generales del inventario."""
    print("\nESTADÍSTICAS DEL INVENTARIO")
    print("=" * 35)

    total_juegos = len(inventario_juegos)
    total_unidades = 0
    suma_valoraciones = 0
    mejor_valorado = None
    mejor_valoracion = 0

    # Recorremos el inventario para acumular datos
    for titulo, datos in inventario_juegos.items():
        total_unidades += datos["stock"]
        suma_valoraciones += datos["valoración"]

        if datos["valoración"] > mejor_valoracion:
            mejor_valoracion = datos["valoración"]
            mejor_valorado = titulo

    promedio_valoracion = suma_valoraciones / total_juegos

    print(f"Títulos distintos: {total_juegos}")
    print(f"Unidades totales en stock: {total_unidades}")
    print(f"Valoración promedio: {promedio_valoracion:.2f}/5")
    print(f"Mejor valorado: {mejor_valorado} ({mejor_valoracion}/5)")
    print(f"Valor total del inventario: ${calcular_valor_total():.2f}")


# 3. Ejecución del programa
mostrar_inventario()
filtrar_por_genero("RPG")
juegos_en_oferta(30.00)
mostrar_estadisticas()
```

# **4. Iterando sobre Listas: Explorando Colecciones 📝**

Cuando trabajamos con listas en Python, podemos recorrer sus elementos de manera secuencial. Es como tener una fila de personas y saludar a cada una mientras avanzamos.

**Características importantes:**

* Las listas mantienen un orden específico.
* Podemos acceder a los elementos por su posición (índice).
* Podemos modificar elementos mientras iteramos.
* Podemos usar el índice, el valor, o ambos.

Imagina que eres profesor/a y debes pasar lista en una clase:
* Tienes una lista de estudiantes (la lista en Python).
* Nombras a cada estudiante en orden (iteración).
* Marcas su asistencia (operación con cada elemento).
* Continúas hasta terminar la lista

💻 Observa el siguiente ejemplo:

```python
# Lista de estudiantes con sus estados
estudiantes = [
    "🎓 Ana (presente)",
    "🎓 Juan (ausente)",
    "🎓 María (presente)",
    "🎓 Carlos (presente)"
]

print("📋 Control de Asistencia")
print("=" * 30)

# Método 1: Iteración simple
print("\nMétodo 1: Lista simple")
for estudiante in estudiantes:
    print(estudiante)

# Método 2: Con índices
print("\nMétodo 2: Con numeración")
for i in range(len(estudiantes)):
    print(f"{i+1}. {estudiantes[i]}")

# Método 3: Usando enumerate
print("\nMétodo 3: Usando enumerate")
for numero, estudiante in enumerate(estudiantes, 1):
    print(f"Estudiante #{numero}: {estudiante}")
```

# **5. Iterando sobre Diccionarios: Navegando Datos Estructurados 🗂️**

Como ya aprendimos, los diccionarios en Python son como un libro organizado por temas: cada tema (clave) tiene su contenido (valor) correspondiente. Cuando iteramos sobre diccionarios, podemos hacerlo de tres formas:

1. Solo por claves (como revisar el índice del libro).
2. Solo por valores (como leer todo el contenido).
3. Por pares clave-valor (como leer cada tema con su contenido).

**Características importantes:**

* Las claves son únicas.
* Podemos acceder directamente a cualquier valor usando su clave.
* Los diccionarios son muy útiles para datos estructurados.

Imagina una agenda telefónica:

* Los nombres son las claves.
* Los números de teléfono son los valores.

Puedes:
* Ver solo los nombres.
* Ver solo los números.
* Ver nombres con sus números correspondientes.

💻 Veamos el siguiente ejemplo:

```python
# Agenda telefónica
contactos = {
    "Ana": "📱 123-456-789",
    "Juan": "📱 987-654-321",
    "María": "📱 555-555-555"
}

print("📒 MI AGENDA DE CONTACTOS")
print("=" * 30)

# Método 1: Iterando sobre claves
print("\n📝 Lista de nombres:")
for nombre in contactos.keys():
    print(f"- {nombre}")

# Método 2: Iterando sobre valores
print("\n📞 Lista de números:")
for numero in contactos.values():
    print(f"- {numero}")

# Método 3: Iterando sobre pares clave-valor
print("\n📋 Directorio completo:")
for nombre, numero in contactos.items():
    print(f"{nombre}: {numero}")
```

2. Creamos una función para mostrar el inventario:

```python
def mostrar_inventario():
    print("\n🎒 INVENTARIO DEL HÉROE")
    print("=" * 40)

    peso_total = 0
    valor_total = 0

    for item, detalles in inventario.items():
        cantidad, peso, valor = detalles
        peso_item = peso * cantidad
        valor_item = valor * cantidad

        print(f"\n{item}:")
        print(f"Cantidad: {cantidad}")
        print(f"Peso: {peso_item}kg")
        print(f"Valor: {valor_item} monedas")

        peso_total += peso_item
        valor_total += valor_item

    print("\n📊 RESUMEN:")
    print(f"Peso total: {peso_total}kg")
    print(f"Valor total: {valor_total} monedas")
```

3. Agregamos función para agregar/quitar items:

```python
def modificar_cantidad(item, delta):
    if item in inventario:
        nueva_cantidad = inventario[item][0] + delta
        if nueva_cantidad >= 0:
            inventario[item][0] = nueva_cantidad
            print(f"\n✅ Cantidad actualizada para: {item}")
        else:
            print("\n❌ No hay suficientes unidades")
    else:
        print("\n❌ Item no encontrado")
```

# **6. La Función range(): Control Numérico Preciso 🔢**

La función range() es una herramienta especial en Python que genera una secuencia de números. Es como tener una regla numérica que podemos ajustar según necesitemos. Puede usarse de tres formas:

1. range(stop): Genera números desde 0 hasta stop-1
2. range(start, stop): Genera números desde start hasta stop-1
3. range(start, stop, step): Genera números desde start hasta stop-1, saltando de step en step

**Características importantes:**

* Es eficiente en memoria.
* Genera números bajo demanda.
* Es perfecto para bucles controlados numéricamente.
* No genera una lista real hasta que se necesita.

Imagina diferentes formas de contar:

* Contar del 1 al 10 (range(1, 11))
* Contar de 2 en 2 hasta 10 (range(0, 11, 2))
* Cuenta regresiva desde 5 (range(5, 0, -1))

Es como cuando un árbitro cuenta:

* Normal: "1, 2, 3, 4, 5"
* De dos en dos: "2, 4, 6, 8, 10"
* Regresivo: "5, 4, 3, 2, 1"

💻 Veamos el siguiente ejemplo:

```python
print("🔢 Explorando range()")
print("=" * 30)

# Ejemplo 1: range(stop)
print("\nContando hasta 5:")
for i in range(5):
    print(f"Número: {i} {'⭐' * i}")

# Ejemplo 2: range(start, stop)
print("\nContando del 5 al 8:")
for i in range(5, 9):
    print(f"Número: {i} {'🌟' * (i-4)}")

# Ejemplo 3: range(start, stop, step)
print("\nContando de 2 en 2:")
for i in range(0, 10, 2):
    print(f"Número: {i} {'✨' * (i//2)}")

# Ejemplo 4: Cuenta regresiva
print("\nCuenta regresiva:")
for i in range(5, -1, -1):
    print(f"Tiempo restante: {i} {'🚀' * i}")
```

# **🎮 Ejercicio Guiado: Sistema de Niveles de Juego**

Objetivo: Crear un sistema que muestre el progreso de niveles en un juego, con diferentes recompensas y desafíos.

Instrucciones:

1. Primero, configuramos las estructuras básicas:

```python
def generar_recompensa(nivel):
    recompensas = {
        "monedas": nivel * 100,
        "experiencia": nivel * 150,
        "items": nivel // 3
    }
    return recompensas

def mostrar_nivel(nivel, recompensas):
    print(f"\n🎮 NIVEL {nivel}")
    print("=" * 20)
    print(f"💰 Monedas: {recompensas['monedas']}")
    print(f"⭐ Experiencia: {recompensas['experiencia']}")
    print(f"🎁 Items: {recompensas['items']}")
```

2. Implementamos el sistema de progresión:

```python
def mostrar_progresion():
    print("🏆 PROGRESIÓN DE NIVELES")
    print("=" * 30)

    for nivel in range(1, 11):
        recompensas = generar_recompensa(nivel)
        mostrar_nivel(nivel, recompensas)

        # Mostrar barra de progreso
        progreso = "▮" * nivel + "▯" * (10-nivel)
        print(f"\nProgreso: [{progreso}] {nivel*10}%")

        if nivel < 10:
            input("\nPresiona Enter para el siguiente nivel...")

mostrar_progresion()
```

# **🎯 ¡Manos a la obra! - Tabla de Multiplicar Interactiva**

Objetivo: Crear un sistema interactivo para practicar las tablas de multiplicar que incluya:

* Diferentes niveles de dificultad.
* Sistema de puntuación.
* Progresión de dificultad.
* Retroalimentación visual.

Instrucciones:

1. Implementa las siguientes funciones:
 * Generar ejercicios según nivel.
 * Verificar respuestas.
 * Mostrar puntuación.
 * Mostrar progreso.
2. Utiliza range() para:
 * Controlar la dificultad.
 * Generar números aleatorios apropiados.
 * Manejar la progresión.

**Tu código aquí:**

```python
import random

# ---------------------------------------------------------------
# Configuración de niveles.
# Usamos range() para CONTROLAR LA DIFICULTAD: define qué tablas y
# qué multiplicadores pueden aparecer en cada nivel.
# ---------------------------------------------------------------
NIVELES = {
    1: {"nombre": "🟢 FÁCIL",   "rango": range(1, 6)},    # tablas del 1 al 5
    2: {"nombre": "🟡 MEDIO",   "rango": range(1, 10)},   # tablas del 1 al 9
    3: {"nombre": "🔴 DIFÍCIL", "rango": range(1, 13)},   # tablas del 1 al 12
}
NIVEL_MAX = 3


def generar_ejercicio(nivel):
    """Genera dos números (a, b) aleatorios apropiados para el nivel.
    random.choice elige sobre los valores del range() del nivel."""
    valores = list(NIVELES[nivel]["rango"])
    a = random.choice(valores)
    b = random.choice(valores)
    return a, b


def verificar_respuesta(a, b, respuesta):
    """Devuelve True si la respuesta coincide con el producto."""
    return respuesta == a * b


def mostrar_puntuacion(puntos, total):
    """Muestra la puntuación final y el porcentaje de aciertos."""
    print("\n" + "=" * 40)
    print(f"📊 PUNTUACIÓN FINAL: {puntos}/{total}")
    if total > 0:
        porcentaje = puntos / total * 100
        print(f"🎯 Precisión: {porcentaje:.1f}%")
        if porcentaje == 100:
            print("🏆 ¡PERFECTO! Dominas las tablas.")
        elif porcentaje >= 70:
            print("👍 ¡Muy bien! Sigue practicando.")
        else:
            print("💪 ¡Ánimo! La práctica hace al maestro.")
    print("=" * 40)


def mostrar_progreso(actual, total):
    """Dibuja una barra de progreso visual."""
    barra = "▮" * actual + "▯" * (total - actual)
    porcentaje = int(actual / total * 100)
    print(f"\nProgreso: [{barra}] {porcentaje}%")


def practicar_tablas(total_ejercicios=10):
    print("🔢 PRÁCTICA DE TABLAS DE MULTIPLICAR")
    print("=" * 40)
    print("\nResponde cada multiplicación. ✨")
    print("• 3 aciertos seguidos = subes de nivel 🚀")
    print("• 1 error = bajas de nivel ⬇️")
    print("• Escribe 'salir' para terminar antes.\n")

    puntos = 0
    respondidos = 0
    nivel = 1
    racha = 0

    # range() MANEJA LA PROGRESIÓN: controla cuántos ejercicios habrá.
    for _ in range(1, total_ejercicios + 1):
        nombre = NIVELES[nivel]["nombre"]
        print(f"\n--- Ejercicio {respondidos + 1} | Nivel {nivel}: {nombre} ---")

        a, b = generar_ejercicio(nivel)
        entrada = input(f"❓ ¿Cuánto es {a} x {b}?  ")

        #strip elimina espacio al inicio y al final del string
        if entrada.strip().lower() == "salir":
            print("\n👋 Terminaste la práctica antes de tiempo.")
            break

        # Validamos que la entrada sea un número entero.
        if not entrada.strip().lstrip("-").isdigit(): #.lstrip("-"), elimina el caracter "-" solo si esta al inicio del string
            print("⚠️  Escribe un número válido. (Este intento no cuenta)")
            continue

        respuesta = int(entrada)
        respondidos += 1

        if verificar_respuesta(a, b, respuesta):
            puntos += 1
            racha += 1
            print(f"✅ ¡Correcto!  Racha: {racha} 🔥")
```

# **7. Iterando con la Función range(): Combinando Conceptos 🔄**

Cuando combinamos range() con estructuras iterativas, creamos una herramienta poderosa para controlar exactamente cuántas veces y cómo queremos que se ejecute nuestro código. Es especialmente útil cuando:

* Necesitamos un contador preciso.
* Queremos iterar sobre índices.
* Necesitamos generar secuencias numéricas específicas.

**Patrones comunes:**

* Iteración por índices.
* Iteración inversa.
* Iteración con pasos personalizados.
* Combinación con listas y diccionarios.

Piensa en un entrenador deportivo:
* "Haz 10 flexiones" (range(10))
* "Cuenta de 2 en 2 hasta 20" (range(0, 21, 2))
* "Cuenta regresiva desde 5" (range(5, -1, -1))

💻 Veamos el siguiente ejemplo:

```python
print("🏃‍♂️ Entrenador Personal Virtual")
print("=" * 30)

# Simulación de rutina de ejercicios
def rutina_ejercicio():
    # Calentamiento: cuenta regresiva
    print("\n🔥 Calentamiento:")
    for i in range(5, 0, -1):
        print(f"Preparación: {i} {'🕐' * i}")

    # Ejercicios principales
    ejercicios = ["🏃‍♂️ Correr", "💪 Flexiones", "🚲 Bicicleta", "🧘‍♂️ Estirar"]

    for ronda in range(1, 4):  # 3 rondas
        print(f"\n🔄 Ronda {ronda}:")
        for i, ejercicio in enumerate(ejercicios, 1):
            print(f"Ejercicio {i}: {ejercicio}")
            for segundos in range(5, 0, -1):
                print(f"Tiempo restante: {segundos}s")

    # Enfriamiento
    print("\n❄️ Enfriamiento:")
    for i in range(3):
        print(f"Respiración profunda {i+1}: {'😌' * (i+1)}")

rutina_ejercicio()
```

# **🎮 Ejercicio Guiado: Sistema de Puntuación por Niveles**

Objetivo: Crear un sistema que calcule y muestre puntuaciones para diferentes niveles de un juego.

Instrucciones:

1. Primero, configuramos las estructuras base:

```python
def calcular_puntos(nivel, dificultad):
    puntos_base = nivel * 100
    multiplicador = {
        "fácil": 1,
        "normal": 1.5,
        "difícil": 2
    }
    return int(puntos_base * multiplicador[dificultad])

def mostrar_tabla_puntos():
    dificultades = ["fácil", "normal", "difícil"]

    print("\n🏆 TABLA DE PUNTUACIONES POR NIVEL")
    print("=" * 40)

    # Encabezado
    print("\nNivel | ", end="")
    for dif in dificultades:
        print(f"{dif.capitalize():^10}", end=" | ")
    print("\n" + "-" * 40)

    # Puntuaciones por nivel
    for nivel in range(1, 11):
        print(f"{nivel:^5} | ", end="")
        for dif in dificultades:
            puntos = calcular_puntos(nivel, dif)
            print(f"{puntos:^10}", end=" | ")
        print()

mostrar_tabla_puntos()
```

2. Agregamos funcionalidad de progreso:

```python
def mostrar_progreso_nivel(nivel, experiencia, exp_necesaria):
    print(f"\n📊 Progreso Nivel {nivel}")
    porcentaje = (experiencia / exp_necesaria) * 100
    barras = int(porcentaje / 5)
    print(f"[{'█' * barras}{'░' * (20-barras)}] {porcentaje:.1f}%")
```
