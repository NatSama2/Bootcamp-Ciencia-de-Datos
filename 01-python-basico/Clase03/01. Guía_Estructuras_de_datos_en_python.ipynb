## **2.1 ¿Qué son las listas? 📋**
Una lista en Python es como una cadena de vagones de tren 🚂:

* Cada vagón puede llevar cualquier tipo de carga (datos).

* Ordenada: Como los vagones del tren, cada elemento tiene una posición específica.
* Mutable: Puedes modificar, agregar o eliminar elementos.
* Indexada: Cada elemento tiene un número de índice (comenzando desde 0).

# **2.3 Agregar Elementos a una Lista 📥**
Hay varias formas de agregar elementos a una lista, igual que hay diferentes formas de agregar vagones a un tren 🚂:

```python
# 1. Usando append() - Agrega al final de la lista
personajes = ["Mario"]
personajes.append("Luigi")  # ["Mario", "Luigi"]

# 2. Usando insert() - Agrega en una posición específica
personajes.insert(1, "Yoshi")  # ["Mario", "Yoshi", "Luigi"]

# 3. Usando extend() - Agrega varios elementos al final
más_personajes = ["Peach", "Bowser"]
personajes.extend(más_personajes)  # ["Mario", "Yoshi", "Luigi", "Peach", "Bowser"]
```

# **💻 Ejercicio Guiado: Juego del Tres en Raya**

```python
# Crear un tablero de tres en raya
def crear_tablero():
    return [
        [' ', ' ', ' '],
        [' ', ' ', ' '],
        [' ', ' ', ' ']
    ]

# Mostrar el tablero
def mostrar_tablero(tablero):
    print('\n  0 1 2')
    for i, fila in enumerate(tablero):
        print(f"{i} {fila[0]}|{fila[1]}|{fila[2]}")
        if i < 2:
            print("  -+-+-")

# Crear y mostrar un tablero vacío
tablero = crear_tablero()
mostrar_tablero(tablero)

# Hacer un movimiento
tablero[1][1] = '❌'
mostrar_tablero(tablero)
```

# **🎯 ¡Manos a la obra! Red Social Simplificada**

Objetivo: Crear un sistema básico de perfiles de usuario para una red social.

Instrucciones:

1. Crea un diccionario para almacenar perfiles de usuarios.
2. Cada perfil debe contener:
  * Nombre de usuario
  * Edad
  * Lista de amigos
  * Estado (online/offline)
  * Publicaciones recientes

3. Implementa funciones para:
  * Agregar nuevo usuario
  * Actualizar estado
  * Agregar amigos
  * Hacer una publicación
  * Mostrar perfil completo



**Tu código aquí:**

```python
# Escribe tu solución aquí
red_social = {
    "usuarios": {}
}

# Agregar nuevo usuario
def agregar_usuario(nombre_usuario, edad):
    red_social["usuarios"][nombre_usuario] = {
        "edad": edad,
        "amigos": [],
        "estado": "offline",
        "publicaciones": []
    }

# Actualizar estado
def actualizar_estado(nombre_usuario, estado):
    if nombre_usuario in red_social["usuarios"]:
        red_social["usuarios"][nombre_usuario]["estado"] = estado

# Agregar amigos
def agregar_amigo(usuario, amigo):
    if usuario in red_social["usuarios"] and amigo in red_social["usuarios"]:#con eto valido que tanto el usuario como la persona que sera amigo exista en resde solciales
        if amigo not in red_social["usuarios"][usuario]["amigos"]:#ei la persona que es amigo no es parte de sus amigo, la agrego
            red_social["usuarios"][usuario]["amigos"].append(amigo)

# Hacer una publicación
def hacer_publicacion(nombre_usuario, publicacion):
    if nombre_usuario in red_social["usuarios"]:
        red_social["usuarios"][nombre_usuario]["publicaciones"].append(publicacion)

# Mostrar perfil completo
def mostrar_perfil(nombre_usuario):
    if nombre_usuario in red_social["usuarios"]:
        perfil = red_social["usuarios"][nombre_usuario]
        print(f"Usuario: {nombre_usuario}")
        print(f"Edad: {perfil['edad']}")
        print(f"Amigos: {perfil['amigos']}")
        print(f"Estado: {perfil['estado']}")
        print(f"Publicaciones: {perfil['publicaciones']}")
    else:
        print("Usuario no encontrado")


# Ejemplo de uso
agregar_usuario("Ana", 20)
agregar_usuario("Luis", 22)

actualizar_estado("Ana", "online")
agregar_amigo("Ana", "Luis")

hacer_publicacion("Ana", "Hola, esta es mi primera publicación")
hacer_publicacion("Ana", "Aprendiendo Python")

mostrar_perfil("Ana")
```

# **💻 Ejercicio Guiado: Sistema de Registro de Estudiantes 👨‍🎓**

```python
def crear_registro_estudiante():
    # Pedimos datos al usuario
    nombre = input("Nombre del estudiante: ")
    edad = int(input("Edad: "))
    notas = tuple(float(input(f"Nota {i+1}: ")) for i in range(3))

    # Creamos una tupla con toda la información
    estudiante = (nombre, edad, notas)
    return estudiante

def mostrar_info_estudiante(estudiante):
    nombre, edad, notas = estudiante  # Desempaquetado
    promedio = sum(notas) / len(notas)

    print("\n📊 Información del Estudiante")
    print("=" * 30)
    print(f"📝 Nombre: {nombre}")
    print(f"🔢 Edad: {edad}")
    print(f"📚 Notas: {notas}")
    print(f"📈 Promedio: {promedio:.2f}")

# Probemos el sistema
estudiante1 = crear_registro_estudiante()
mostrar_info_estudiante(estudiante1)
```

  * Calcular la distancia entre dos puntos.
  * Determinar si tres puntos forman un triángulo.
  * Calcular el punto medio entre dos puntos.



**Tu código aquí:**

```python
# Escribe tu solución aquí
def crear_punto_2d(x, y):
    return (x, y)

def crear_punto_3d(x, y, z):
    return (x, y, z)

# Agrega más funciones...
# 1. Función para crear el punto 2D
def crear_punto_2d(x, y):
    return (x, y)

# 2. Función para calcular la distancia (Sin importar math)
def calcular_distancia(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    
    # Todo el paréntesis se eleva a 0.5 para sacar la raíz cuadrada
    return ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5

# --------------------------------------------------------
# 🎯 PRUEBA
# --------------------------------------------------------
punto_A = crear_punto_2d(0, 0)
punto_B = crear_punto_2d(3, 4)

print(f"Punto A: {punto_A}")
print(f"Punto B: {punto_B}")
print(f"Distancia: {calcular_distancia(punto_A, punto_B)}")
```

# **5. Sets: ¡Colecciones Únicas y Poderosas! 🎯**
# **5.1 ¿Qué es un Set? 🤔**

Imagina una colección de estampillas: aunque alguien te regale una estampilla que ya tienes, no la agregas nuevamente a tu colección. ¡Así funcionan los sets! Son colecciones de elementos únicos, sin duplicados.

```python
# Ejemplo de un set
colores_primarios = {"rojo", "azul", "amarillo"}

# Intentando agregar un duplicado (no tendrá efecto)
colores_primarios.add("rojo")
print(colores_primarios)  # Solo mostrará: {'rojo', 'azul', 'amarillo'}
```

# **5.2 Características de un Set ⭐**

* Elementos únicos: No permite duplicados.
* Desordenado: No mantiene un orden específico.
* Mutable: Podemos agregar o eliminar elementos.
* Elementos inmutables: Solo puede contener elementos inmutables (números, strings, tuplas).

# **5.4 Operaciones de Conjunto con sets 🎲**

Los sets son perfectos para operaciones matemáticas de conjuntos:

```python
# Definimos dos grupos de estudiantes
python = {"Ana", "Juan", "María", "Carlos"}
javascript = {"María", "Pedro", "Juan", "Isabel"}

# Unión (estudiantes en cualquier curso)
todos = python.union(javascript)
# o también: python | javascript

# Intersección (estudiantes en ambos cursos)
ambos = python.intersection(javascript)
# o también: python & javascript

# Diferencia (estudiantes solo en Python)
solo_python = python.difference(javascript)
# o también: python - javascript

# Diferencia simétrica (estudiantes en un curso pero no en ambos)
uno_u_otro = python.symmetric_difference(javascript)
# o también: python ^ javascript
```

# **💻 Ejercicio Guiado: Sistema de Gestión de Biblioteca 📚**

```python
class BibliotecaDigital:
    def __init__(self):
        self.todos_libros = set()         # Todos los libros
        self.libros_prestados = set()     # Libros actualmente prestados
        self.libros_disponibles = set()   # Libros disponibles

    def agregar_libro(self, libro):
        self.todos_libros.add(libro)
        self.libros_disponibles.add(libro)
        print(f"📚 Libro '{libro}' agregado a la biblioteca")

    def prestar_libro(self, libro):
        if libro in self.libros_disponibles:
            self.libros_disponibles.remove(libro)
            self.libros_prestados.add(libro)
            print(f"✅ Libro '{libro}' prestado exitosamente")
        else:
            print(f"❌ Libro '{libro}' no disponible")

    def devolver_libro(self, libro):
        if libro in self.libros_prestados:
            self.libros_prestados.remove(libro)
            self.libros_disponibles.add(libro)
            print(f"✅ Libro '{libro}' devuelto exitosamente")
        else:
            print(f"❌ El libro '{libro}' no estaba prestado")

    def mostrar_estado(self):
        print("\n📊 Estado de la Biblioteca")
        print("=" * 30)
        print(f"📚 Total de libros: {len(self.todos_libros)}")
        print(f"📖 Libros disponibles: {len(self.libros_disponibles)}")
        print(f"📘 Libros prestados: {len(self.libros_prestados)}")

# Probemos el sistema
biblioteca = BibliotecaDigital()
biblioteca.agregar_libro("Python para Principiantes")
biblioteca.agregar_libro("Aprende Sets en Python")
biblioteca.agregar_libro("Estructuras de Datos")

biblioteca.prestar_libro("Python para Principiantes")
biblioteca.mostrar_estado()
```
