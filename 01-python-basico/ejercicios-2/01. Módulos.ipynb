## Explorando el Módulo Statistics 📊

El módulo statistics nos ayuda a analizar conjuntos de datos. Es como tener un asistente que nos ayuda a entender números.

Funciones principales:
- mean(): calcula el promedio
- median(): encuentra el valor del medio
- mode(): encuentra el valor que más se repite

````python
# Imaginemos que estas son las notas de un estudiante
notas = [5.5, 6.0, 6.0, 5.0, 6.5, 6.0, 4.5]

print("=== Análisis de Notas ===")
print(f"Notas del estudiante: {notas}")
print(f"Promedio: {statistics.mean(notas):.1f}")
print(f"Nota del medio: {statistics.median(notas)}")
print(f"Nota más común: {statistics.mode(notas)}")
````

### 🎯 ¡Manos a la obra!: Registro de Temperaturas

Imaginemos que trabajamos en una estación meteorológica y necesitamos analizar las temperaturas de la semana.

````python
# Temperaturas de la semana (en grados Celsius)
temperaturas = [24, 25, 23, 27, 25, 22, 24]

# Análisis básico
print("=== Análisis de Temperaturas ===")
print(f"Temperaturas registradas: {temperaturas}°C")
print(f"Temperatura promedio: {statistics.mean(temperaturas)}°C")
print(f"Temperatura más frecuente: {statistics.mode(temperaturas)}°C")

# Usando math para cálculos adicionales
temp_promedio = statistics.mean(temperaturas)
variacion = math.sqrt(statistics.variance(temperaturas))

print(f"\nVariación de temperatura: ±{variacion:.1f}°C")
print(f"Temperatura máxima esperada: {temp_promedio + variacion:.1f}°C")
print(f"Temperatura mínima esperada: {temp_promedio - variacion:.1f}°C")
````

### 🏃 ¡Manos a la obra!: Análisis de Rendimiento Deportivo

Eres un entrenador personal y necesitas analizar los tiempos de carrera de un atleta.

````python
def analizar_rendimiento():
    # Tiempos de carrera en minutos
    tiempos = [25.5, 24.8, 26.0, 23.9, 24.5]

    print("=== Análisis de Rendimiento ===")
    print(f"Tiempos registrados: {tiempos} minutos")

    # Análisis estadístico
    tiempo_promedio = statistics.mean(tiempos)
    tiempo_mejor = min(tiempos)

    print(f"\nTiempo promedio: {tiempo_promedio:.1f} minutos")
    print(f"Mejor tiempo: {tiempo_mejor} minutos")

    # Calculando mejora potencial
    diferencia = tiempo_promedio - tiempo_mejor
    mejora_potencial = math.sqrt(diferencia) * 10

    print(f"\nPotencial de mejora: {mejora_potencial:.1f}%")

# Ejecutamos el análisis
analizar_rendimiento()
````

### 🏪 ¡Manos a la obra!: Sistema de Análisis de Ventas

Vamos a crear un sistema para una tienda que necesita analizar sus ventas diarias.

````python
def analizar_ventas():
    print("=== SISTEMA DE ANÁLISIS DE VENTAS ===")

    # Simulamos ventas diarias de una semana
    ventas = []
    for dia in range(7):
        venta = float(input(f"Ingrese ventas del día {dia+1}: $"))
        ventas.append(venta)

    # Análisis estadístico
    venta_promedio = statistics.mean(ventas)
    venta_mediana = statistics.median(ventas)

    print("\n=== Resultados ===")
    print(f"Ventas totales: ${sum(ventas)}")
    print(f"Venta promedio: ${venta_promedio:.2f}")
    print(f"Venta mediana: ${venta_mediana:.2f}")

    # Análisis adicional usando math
    desviacion = math.sqrt(statistics.variance(ventas))
    print(f"\nVariación en ventas: ±${desviacion:.2f}")

    # Proyección
    proyeccion_mensual = venta_promedio * 30
    print(f"\nProyección mensual: ${proyeccion_mensual:.2f}")

    # Intervalo de confianza
    margen = math.sqrt(desviacion) * 2
    print(f"Rango esperado de ventas diarias:")
    print(f"Mínimo: ${venta_promedio - margen:.2f}")
    print(f"Máximo: ${venta_promedio + margen:.2f}")

# Ejecutamos el análisis
analizar_ventas()
````
