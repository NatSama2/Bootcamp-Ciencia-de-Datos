## 2. Funciones con Parámetros

Los parámetros son información que le pasamos a una función para que trabaje con ella.
Crearemos una función que saluda al usuario por su nombre.

```python
def saludar_creador(nombre):
    print(f"¡Hola {nombre}!")
    print("Juntos crearemos una historia increíble")

# Probamos la función
nombre_autor = input("¿Cómo te llamas? ")
saludar_creador(nombre_autor)
```

### Función con Múltiples Parámetros
Ahora crearemos una función que usa varios parámetros para crear un personaje.

```python
def crear_personaje(nombre, tipo):
    print("=== TU PERSONAJE ===")
    print(f"Nombre: {nombre}")
    if tipo == "dragón":
        print("Tipo: Dragón")
    elif tipo == "hada":
        print("Tipo: Hada")
    elif tipo == "mago":
        print("Tipo: Mago")
    else:
        print(f"Tipo: {tipo}")

# Probemos la función
nombre_personaje = input("Nombre del personaje: ")
tipo_personaje = input("Tipo (dragón/hada/mago/otro): ")
crear_personaje(nombre_personaje, tipo_personaje)
```
