# **📱 Ejercicio guiado - Sistema de Análisis de Reseñas de Productos**

Trabajas para una tienda de comercio electrónico que vende productos tecnológicos. La tienda necesita analizar las reseñas que los clientes dejan sobre los productos para entender mejor la opinión de los usuarios y mejorar su servicio.

La tienda recibe diariamente muchas reseñas de productos y necesita una herramienta que permita:

* Contar cuántas palabras tiene cada reseña.
* Identificar palabras clave que más se repiten.
* Obtener estadísticas básicas de las reseñas.

📋 Ejemplo de Reseñas

```python
reseña1 = """
El smartphone es excelente. La batería dura todo el día y la cámara toma fotos increíbles.
El precio es un poco alto pero vale la pena por la calidad. Muy satisfecho con mi compra.
"""

reseña2 = """
No estoy contento con la compra. La batería dura poco y el sistema es lento.
El servicio al cliente tampoco fue bueno cuando reporté los problemas.
"""
```

Para resolver este problema necesitaremos:

* Crear una clase que procese texto.
* Utilizar métodos para análisis de strings.
* Usar librerías de Python para análisis más avanzado.

📝 Paso 1: Creando nuestra primera clase

Empezaremos creando una clase simple para analizar las reseñas.

```python
class AnalizadorReseñas:
    def __init__(self, reseña):
        """
        Inicializa el analizador con una reseña
        Parámetros:
            reseña (str): El texto de la reseña a analizar
        """
        self.reseña = reseña
        self.palabras = reseña.split()

    def contar_palabras(self):
        """
        Cuenta el número total de palabras en la reseña
        Retorna:
            int: Número de palabras
        """
        return len(self.palabras)

    def contar_caracteres(self):
        """
        Cuenta el número de caracteres en la reseña
        Retorna:
            int: Número de caracteres
        """
        return len(self.reseña)
```

Probemos nuestra clase con una reseña real:

```python
# Ejemplo de uso
reseña_ejemplo = """
El smartphone es excelente. La batería dura todo el día y la cámara toma fotos increíbles.
El precio es un poco alto pero vale la pena por la calidad. Muy satisfecho con mi compra.
"""

# Creamos una instancia de nuestro analizador
analizador = AnalizadorReseñas(reseña_ejemplo)

# Probamos los métodos básicos
print("Análisis básico de la reseña:")
print(f"Número de palabras: {analizador.contar_palabras()}")
print(f"Número de caracteres: {analizador.contar_caracteres()}")
```

📝 Paso 2: Usando librerías de Python

Ahora mejoraremos nuestro analizador utilizando librerías existentes.

```python
from collections import Counter
import string

class AnalizadorReseñasAvanzado:
    def __init__(self, reseña):
        """
        Inicializa el analizador con capacidades avanzadas
        Parámetros:
            reseña (str): El texto de la reseña a analizar
        """
        self.reseña = reseña
        # Limpiamos el texto de puntuación
        self.texto_limpio = reseña.translate(
            str.maketrans("", "", string.punctuation)
        )
        self.palabras = self.texto_limpio.lower().split()

    def palabras_frecuentes(self, n=3):
        """
        Encuentra las palabras más repetidas
        Parámetros:
            n (int): Número de palabras más frecuentes a retornar
        Retorna:
            list: Lista de tuplas (palabra, frecuencia)
        """
        contador = Counter(self.palabras)
        return contador.most_common(n)

    def obtener_estadisticas(self):
        """
        Calcula estadísticas de la reseña
        Retorna:
            dict: Diccionario con estadísticas
        """
        return {
            'total_palabras': len(self.palabras),
            'total_caracteres': len(self.reseña),
            'palabras_unicas': len(set(self.palabras))
        }
```

Probemos el analizador avanzado:

```python
# Creamos instancia del analizador avanzado
analizador_avanzado = AnalizadorReseñasAvanzado(reseña_ejemplo)

# Obtenemos estadísticas
stats = analizador_avanzado.obtener_estadisticas()
print("\nEstadísticas de la reseña:")
for key, value in stats.items():
    print(f"{key}: {value}")

# Vemos las palabras más frecuentes
print("\nPalabras más frecuentes:")
for palabra, frecuencia in analizador_avanzado.palabras_frecuentes():
    print(f"'{palabra}' aparece {frecuencia} veces")
```

```python
# Primero importamos las librerías necesarias
from collections import Counter
import string

# Definimos la primera clase
class AnalizadorReseñas:
    def __init__(self, reseña):
        """
        Inicializa el analizador con una reseña
        Parámetros:
            reseña (str): El texto de la reseña a analizar
        """
        self.reseña = reseña
        self.palabras = reseña.split()

    def contar_palabras(self):
        """
        Cuenta el número total de palabras en la reseña
        Retorna:
            int: Número de palabras
        """
        return len(self.palabras)

    def contar_caracteres(self):
        """
        Cuenta el número de caracteres en la reseña
        Retorna:
            int: Número de caracteres
        """
        return len(self.reseña)

# Definimos la segunda clase
class AnalizadorReseñasAvanzado:
    def __init__(self, reseña):
        """
        Inicializa el analizador con capacidades avanzadas
        Parámetros:
            reseña (str): El texto de la reseña a analizar
        """
        self.reseña = reseña
        # Limpiamos el texto de puntuación
        self.texto_limpio = reseña.translate(
            str.maketrans("", "", string.punctuation)
        )
        self.palabras = self.texto_limpio.lower().split()

    def palabras_frecuentes(self, n=3):
        """
        Encuentra las palabras más repetidas
        Parámetros:
            n (int): Número de palabras más frecuentes a retornar
        Retorna:
            list: Lista de tuplas (palabra, frecuencia)
        """
        contador = Counter(self.palabras)
        return contador.most_common(n)

    def obtener_estadisticas(self):
        """
        Calcula estadísticas de la reseña
        Retorna:
            dict: Diccionario con estadísticas
        """
        return {
            'total_palabras': len(self.palabras),
            'total_caracteres': len(self.reseña),
            'palabras_unicas': len(set(self.palabras))
        }

# Lista de reseñas para analizar
reseñas = [
    """El smartphone es excelente. La batería dura todo el día y la cámara es increíble.""",
    """No estoy contento con la compra. La batería dura poco y el sistema es lento.""",
    """Buen producto por el precio. Cumple con lo prometido."""
]

# Analizamos cada reseña
for i, reseña in enumerate(reseñas, 1):
    print(f"\n--- Análisis de Reseña {i} ---")

    # Usamos ambos analizadores
    analizador_basico = AnalizadorReseñas(reseña)
    analizador_avanzado = AnalizadorReseñasAvanzado(reseña)

    # Mostramos resultados básicos
    print(f"Palabras: {analizador_basico.contar_palabras()}")
    print(f"Caracteres: {analizador_basico.contar_caracteres()}")

    # Mostramos palabras frecuentes
    print("\nPalabras más usadas:")
    for palabra, freq in analizador_avanzado.palabras_frecuentes(2):
        print(f"- {palabra}: {freq} veces")
```
